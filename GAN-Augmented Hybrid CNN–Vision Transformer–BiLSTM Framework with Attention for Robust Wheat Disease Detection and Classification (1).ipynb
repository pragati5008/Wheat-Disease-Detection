```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
import numpy as np
from PIL import Image
import os
import random
import torchvision.transforms as transforms
import timm
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, \
    precision_recall_fscore_support, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import warnings
from torch.optim.lr_scheduler import CosineAnnealingLR
import albumentations as A
from albumentations.pytorch import ToTensorV2
from sklearn.preprocessing import label_binarize

warnings.filterwarnings("ignore")

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
scaler = torch.cuda.amp.GradScaler() if device.type == 'cuda' else None
print(f"Using device: {device}")

class Config:
    IMAGE_SIZE = 224
    BATCH_SIZE = 64
    NUM_WORKERS = 8
    NUM_CLASSES = 5

    # Reduced epochs for classifier
    EPOCHS = 15
    LEARNING_RATE = 3e-4
    WEIGHT_DECAY = 1e-4

    # Reduced GAN epochs
    GAN_EPOCHS = 20
    GAN_BATCH_SIZE = 32
    GAN_LR = 2e-4
    GAN_FEATURE_MATCHING = True
    GAN_AUGMENTATION_RATIO = 1

    TRAIN_SPLIT_RATIO = 0.8
    TEST_SPLIT_RATIO = 0.1
    VIT_BACKBONE = 'vit_base_patch16_224'
    CNN_BACKBONE = 'efficientnet_b2'
    LSTM_HIDDEN_SIZE = 256
    LSTM_NUM_LAYERS = 2
    LSTM_DROPOUT = 0.3
    SEQUENCE_LENGTH = 6

# ===================== GAN MODULES =====================

class ConditionalGenerator(nn.Module):
    def __init__(self, num_classes, latent_dim=100, img_channels=3):
        super().__init__()
        self.label_emb = nn.Embedding(num_classes, 64)

        self.model = nn.Sequential(
            nn.Linear(latent_dim + 64, 256 * 4 * 4),
            nn.ReLU(),
            nn.Unflatten(1, (256, 4, 4)),
            nn.ConvTranspose2d(256, 128, 4, 2, 1), nn.BatchNorm2d(128), nn.ReLU(),
            nn.ConvTranspose2d(128, 64, 4, 2, 1), nn.BatchNorm2d(64), nn.ReLU(),
            nn.ConvTranspose2d(64, 32, 4, 2, 1), nn.BatchNorm2d(32), nn.ReLU(),
            nn.ConvTranspose2d(32, img_channels, 4, 2, 1), nn.Tanh()
        )

    def forward(self, z, labels):
        c = self.label_emb(labels)
        x = torch.cat([z, c], dim=1)
        img = self.model(x)
        return img

class ConditionalDiscriminator(nn.Module):
    def __init__(self, num_classes, img_channels=3):
        super().__init__()
        self.label_emb = nn.Embedding(num_classes, 64)

        self.model = nn.Sequential(
            nn.Conv2d(img_channels + 64, 64, 4, 2, 1), nn.LeakyReLU(0.2),
            nn.Conv2d(64, 128, 4, 2, 1), nn.BatchNorm2d(128), nn.LeakyReLU(0.2),
            nn.Conv2d(128, 256, 4, 2, 1), nn.BatchNorm2d(256), nn.LeakyReLU(0.2),
            nn.Conv2d(256, 512, 4, 2, 1), nn.BatchNorm2d(512), nn.LeakyReLU(0.2),
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),
            nn.Linear(512, 1)
        )

    def forward(self, img, labels):
        c = self.label_emb(labels).view(labels.size(0), 64, 1, 1)
        c = c.repeat(1, 1, img.size(2), img.size(3))
        x = torch.cat([img, c], dim=1)
        return self.model(x)

class WheatDiseaseGAN:
    def __init__(self, num_classes):
        self.num_classes = num_classes
        self.generator = ConditionalGenerator(num_classes).to(device)
        self.discriminator = ConditionalDiscriminator(num_classes).to(device)

        self.g_optimizer = optim.Adam(self.generator.parameters(), lr=Config.GAN_LR, betas=(0.5, 0.999))
        self.d_optimizer = optim.Adam(self.discriminator.parameters(), lr=Config.GAN_LR, betas=(0.5, 0.999))

        print(f"🆕 GAN - G: {sum(p.numel() for p in self.generator.parameters())/1e6:.1f}M, "
              f"D: {sum(p.numel() for p in self.discriminator.parameters())/1e6:.1f}M")

    def train_gan(self, dataloader, epochs=Config.GAN_EPOCHS):
        print("🚀 Training Conditional GAN...")
        criterion = nn.BCEWithLogitsLoss()

        for epoch in range(epochs):
            g_loss_epoch, d_loss_epoch = 0, 0

            pbar = tqdm(dataloader, desc=f'GAN E{epoch+1}', leave=False)
            for real_imgs, labels in pbar:
                batch_size = real_imgs.size(0)
                real_imgs = real_imgs.to(device)
                labels = labels.to(device)

                real_labels = torch.ones(batch_size, 1).to(device)
                fake_labels = torch.zeros(batch_size, 1).to(device)

                # === TRAIN DISCRIMINATOR ===
                self.d_optimizer.zero_grad()
                real_pred = self.discriminator(real_imgs, labels)
                d_real_loss = criterion(real_pred, real_labels)

                z = torch.randn(batch_size, 100).to(device)
                fake_imgs = self.generator(z, labels)
                fake_pred = self.discriminator(fake_imgs.detach(), labels)
                d_fake_loss = criterion(fake_pred, fake_labels)

                d_loss = (d_real_loss + d_fake_loss) / 2
                d_loss.backward()
                self.d_optimizer.step()

                # === TRAIN GENERATOR ===
                self.g_optimizer.zero_grad()
                fake_pred = self.discriminator(fake_imgs, labels)
                g_loss = criterion(fake_pred, real_labels)

                if Config.GAN_FEATURE_MATCHING:
                    feat_real = torch.mean(real_imgs, dim=[2, 3], keepdim=True)
                    feat_fake = torch.mean(fake_imgs, dim=[2, 3], keepdim=True)
                    fm_loss = F.mse_loss(feat_fake, feat_real)
                    g_loss += 0.1 * fm_loss

                g_loss.backward()
                self.g_optimizer.step()

                g_loss_epoch += g_loss.item()
                d_loss_epoch += d_loss.item()
                pbar.set_postfix({'D': f'{d_loss.item():.3f}', 'G': f'{g_loss.item():.3f}'})

            if epoch % 5 == 0 or epoch == epochs - 1:
                print(f'GAN E[{epoch+1}/{epochs}] D:{d_loss_epoch/len(dataloader):.3f} '
                      f'G:{g_loss_epoch/len(dataloader):.3f}')

        print("✅ GAN trained!")

    @torch.no_grad()
    def generate_synthetic_data(self, num_samples_per_class):
        print("🎨 Generating synthetic images...")
        synthetic_paths, synthetic_labels = [], []
        os.makedirs("synthetic_data", exist_ok=True)

        self.generator.eval()
        with torch.no_grad():
            for class_idx in range(self.num_classes):
                labels = torch.full((num_samples_per_class,), class_idx,
                                    dtype=torch.long).to(device)
                z = torch.randn(num_samples_per_class, 100).to(device)
                fake_imgs = self.generator(z, labels)

                class_dir = f"synthetic_data/class_{class_idx}"
                os.makedirs(class_dir, exist_ok=True)

                for i, img in enumerate(fake_imgs):
                    img = (img * 0.5 + 0.5).clamp(0, 1)
                    # keep all channels
                    img_pil = transforms.ToPILImage()(img.cpu())
                    path = os.path.join(class_dir, f"syn_{i:04d}.png")
                    img_pil.save(path)
                    synthetic_paths.append(path)
                    synthetic_labels.append(class_idx)

        print(f"✅ Generated {len(synthetic_paths)} images")
        return synthetic_paths, np.array(synthetic_labels)

# ===================== DATA & TRANSFORMS =====================

def detect_dataset_structure(data_dir):
    class_names, image_paths, labels = [], [], []
    print(f"🔍 Scanning: {data_dir}")

    for root, dirs, files in os.walk(data_dir):
        image_files = [f for f in files if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        if not image_files:
            continue

        class_name = os.path.basename(root)
        if class_name not in class_names:
            class_names.append(class_name)
            class_names.sort()

        class_idx = class_names.index(class_name)
        for img_file in image_files:
            image_paths.append(os.path.join(root, img_file))
            labels.append(class_idx)

    print(f"✅ Found {len(class_names)} classes: {class_names}")
    print(f"📊 Total images: {len(image_paths)}")
    return image_paths, np.array(labels), class_names

def get_fast_transforms():
    train_transform = A.Compose([
        A.Resize(Config.IMAGE_SIZE, Config.IMAGE_SIZE),
        A.HorizontalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.RandomBrightnessContrast(p=0.3),
        A.Normalize(mean=[0.485, 0.456, 0.406],
                    std=[0.229, 0.224, 0.225]),
        ToTensorV2()
    ])
    val_test_transform = A.Compose([
        A.Resize(Config.IMAGE_SIZE, Config.IMAGE_SIZE),
        A.Normalize(mean=[0.485, 0.456, 0.406],
                    std=[0.229, 0.224, 0.225]),
        ToTensorV2()
    ])
    return train_transform, val_test_transform

class WheatDiseaseDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = list(image_paths)
        self.labels = np.array(labels)
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        try:
            image = Image.open(self.image_paths[idx]).convert('RGB')
            image = np.array(image)
            label = int(self.labels[idx])
            if self.transform:
                augmented = self.transform(image=image)
                image = augmented['image']
            return image, label
        except:
            dummy = np.full((Config.IMAGE_SIZE, Config.IMAGE_SIZE, 3),
                            128, dtype=np.uint8)
            return torch.tensor(dummy.transpose(2, 0, 1) / 255.0), 0

def prepare_gan_augmented_data_simple(data_dir):
    orig_paths, orig_labels, class_names = detect_dataset_structure(data_dir)
    if len(orig_paths) == 0:
        raise ValueError("No images found!")

    _, val_test_transform = get_fast_transforms()

    gan_dataset = WheatDiseaseDataset(orig_paths, orig_labels, val_test_transform)
    gan_loader = DataLoader(
        gan_dataset,
        batch_size=Config.GAN_BATCH_SIZE,
        shuffle=True,
        num_workers=4,
        pin_memory=True
    )

    gan_trainer = WheatDiseaseGAN(len(class_names))
    gan_trainer.train_gan(gan_loader, epochs=Config.GAN_EPOCHS)

    orig_counts = np.bincount(orig_labels)
    target_per_class = min(int(np.min(orig_counts) * Config.GAN_AUGMENTATION_RATIO), 400)
    syn_paths, syn_labels = gan_trainer.generate_synthetic_data(target_per_class)

    all_paths = orig_paths + syn_paths
    all_labels = np.concatenate([orig_labels, syn_labels])

    print(f"📊 GAN Augmented: {len(orig_paths)} → {len(all_paths)} images")
    return all_paths, all_labels, class_names

# ===================== MODEL (CNN + ViT + LSTM) =====================

class CNNFeatureExtractor(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = timm.create_model(Config.CNN_BACKBONE, pretrained=True, num_classes=0)
        for param in self.backbone.parameters():
            param.requires_grad = False
        # Unfreeze last 2 blocks if exist
        if hasattr(self.backbone, "blocks"):
            for param in self.backbone.blocks[-2:].parameters():
                param.requires_grad = True

        dummy = torch.randn(1, 3, Config.IMAGE_SIZE, Config.IMAGE_SIZE)
        with torch.no_grad():
            feats = self.backbone.forward_features(dummy)
        self.feature_dim = feats.shape[1] if feats.ndim == 4 else feats.shape[-1]
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.proj = nn.Sequential(
            nn.Linear(self.feature_dim, 512),
            nn.GELU(),
            nn.Dropout(0.3)
        )

    def forward(self, x):
        feats = self.backbone.forward_features(x)
        if feats.ndim == 4:
            pooled = self.pool(feats).flatten(1)
        else:
            pooled = feats
        return self.proj(pooled)

class ViTFeatureExtractor(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = timm.create_model(Config.VIT_BACKBONE, pretrained=True, num_classes=0)
        for param in self.backbone.parameters():
            param.requires_grad = False
        if hasattr(self.backbone, "blocks"):
            for param in self.backbone.blocks[-2:].parameters():
                param.requires_grad = True

    def forward(self, x):
        return self.backbone(x)

class FastAttention(nn.Module):
    def __init__(self, d_model, num_heads=8):
        super().__init__()
        self.attn = nn.MultiheadAttention(d_model, num_heads,
                                          batch_first=True, dropout=0.1)

    def forward(self, x):
        attn_out, attn_weights = self.attn(x, x, x)
        return attn_out[:, -1:], attn_weights.mean(dim=1)

class FastHybridClassifier(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.cnn = CNNFeatureExtractor()
        self.vit = ViTFeatureExtractor()
        self.lstm = nn.LSTM(512 + 768, 256, 2,
                            batch_first=True, bidirectional=True, dropout=0.2)
        self.attention = FastAttention(512)
        self.classifier = nn.Sequential(
            nn.LayerNorm(512),
            nn.Dropout(0.4),
            nn.Linear(512, 256),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        cnn_feat = self.cnn(x)
        vit_feat = self.vit(x)
        fused = torch.cat([vit_feat, cnn_feat], dim=1).unsqueeze(1)
        seq = [fused + torch.randn_like(fused) * 0.02
               for _ in range(Config.SEQUENCE_LENGTH)]
        seq = torch.cat(seq, dim=1)
        lstm_out, _ = self.lstm(seq)
        attn_out, attn_w = self.attention(lstm_out)
        return self.classifier(attn_out.squeeze(1)), attn_w

# ===================== METRICS & VISUALIZATION =====================

class MetricsVisualizer:
    def __init__(self, class_names, fold_idx=None):
        self.class_names = class_names
        self.fold_idx = fold_idx

    def _prefix(self, name):
        if self.fold_idx is None:
            return name
        base, ext = os.path.splitext(name)
        return f"fold{self.fold_idx}_{base}{ext}"

    def plot_loss_curve(self, train_losses, val_losses):
        plt.figure(figsize=(10, 6))
        plt.plot(train_losses, label='Train Loss', linewidth=3, color='blue')
        plt.plot(val_losses, label='Val Loss', linewidth=3, color='red')
        plt.title('🗺️ Training & Validation Loss', fontsize=16, fontweight='bold', pad=20)
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        fname = self._prefix('1_loss_curve.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_accuracy_curve(self, train_accs, val_accs):
        plt.figure(figsize=(10, 6))
        plt.plot(train_accs, label='Train Acc', linewidth=3, color='green')
        plt.plot(val_accs, label='Val Acc', linewidth=3, color='orange')
        plt.title('📈 Training & Validation Accuracy', fontsize=16, fontweight='bold', pad=20)
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        fname = self._prefix('2_accuracy_curve.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_confusion_matrix(self, y_test, y_pred):
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=self.class_names, yticklabels=self.class_names)
        plt.title('🔍 Confusion Matrix (Test Set)', fontsize=16, fontweight='bold')
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.tight_layout()
        fname = self._prefix('3_confusion_matrix.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_precision_recall(self, y_test, y_pred):
        precision, recall, _, _ = precision_recall_fscore_support(
            y_test, y_pred, average=None)
        x_pos = np.arange(len(self.class_names))
        width = 0.35

        plt.figure(figsize=(12, 6))
        plt.bar(x_pos - width / 2, precision, width,
                label='Precision', alpha=0.8, color='skyblue')
        plt.bar(x_pos + width / 2, recall, width,
                label='Recall', alpha=0.8, color='salmon')
        plt.title('⚖️ Precision & Recall per Class', fontsize=16, fontweight='bold')
        plt.xlabel('Classes')
        plt.ylabel('Score')
        plt.xticks(x_pos, [name[:12] for name in self.class_names], rotation=45)
        plt.legend()
        plt.ylim(0, 1)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        fname = self._prefix('4_precision_recall.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_roc_curves(self, y_test, y_proba):
        plt.figure(figsize=(12, 10))
        y_bin = label_binarize(y_test, classes=range(len(self.class_names)))
        colors = plt.cm.Set1(np.linspace(0, 1, len(self.class_names)))

        for i, color in enumerate(colors):
            if y_bin[:, i].sum() > 0:
                fpr, tpr, _ = roc_curve(y_bin[:, i], y_proba[:, i])
                roc_auc = auc(fpr, tpr)
                plt.plot(fpr, tpr, color=color, lw=2.5,
                         label=f'{self.class_names[i][:15]} (AUC = {roc_auc:.3f})')

        plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Random Classifier')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate', fontsize=12)
        plt.ylabel('True Positive Rate', fontsize=12)
        plt.title('🎯 Multi-Class ROC Curves (GAN-Augmented Test Set)',
                  fontsize=16, fontweight='bold', pad=20)
        plt.legend(loc="lower right", fontsize=11)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        fname = self._prefix('6_roc_curves.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_metrics_summary(self, y_test, y_pred):
        precision_macro, recall_macro, f1_macro, _ = precision_recall_fscore_support(
            y_test, y_pred, average='macro')
        precision_weighted, recall_weighted, f1_weighted, _ = precision_recall_fscore_support(
            y_test, y_pred, average='weighted')
        final_acc = accuracy_score(y_test, y_pred)

        metrics_data = [
            ['Metric', 'Macro Avg', 'Weighted Avg'],
            ['Accuracy', f'{final_acc:.3f}', f'{final_acc:.3f}'],
            ['Precision', f'{precision_macro:.3f}', f'{precision_weighted:.3f}'],
            ['Recall', f'{recall_macro:.3f}', f'{recall_weighted:.3f}'],
            ['F1-Score', f'{f1_macro:.3f}', f'{f1_weighted:.3f}']
        ]

        plt.figure(figsize=(10, 6))
        table = plt.table(cellText=metrics_data[1:],
                          colLabels=metrics_data[0],
                          cellLoc='center', loc='center')
        table.auto_set_font_size(False)
        table.set_fontsize(12)
        table.scale(1, 2.5)
        plt.title('📊 GAN-Augmented Metrics Summary', fontsize=16,
                  fontweight='bold', pad=20)
        plt.axis('off')
        plt.tight_layout()
        fname = self._prefix('7_metrics_summary.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def print_classification_report(self, y_test, y_pred):
        report = classification_report(y_test, y_pred,
                                       target_names=self.class_names,
                                       digits=4)
        print("\n" + "=" * 80)
        print("📋 GAN-AUGMENTED CLASSIFICATION REPORT")
        print("=" * 80)
        print(report)
        fname = self._prefix('classification_report.txt')
        with open(fname, 'w') as f:
            f.write(report)
        print(f"✅ Saved: {fname}")
        print("=" * 80)

# ===================== CLASSIFIER WRAPPER =====================

class FastWheatClassifier:
    def __init__(self, num_classes=5):
        self.model = FastHybridClassifier(num_classes).to(device)
        print(f"Classifier: {sum(p.numel() for p in self.model.parameters())/1e6:.1f}M params")

        self.optimizer = optim.AdamW(self.model.parameters(),
                                     lr=Config.LEARNING_RATE,
                                     weight_decay=Config.WEIGHT_DECAY)
        self.scheduler = CosineAnnealingLR(self.optimizer, T_max=Config.EPOCHS)
        self.criterion = nn.CrossEntropyLoss(label_smoothing=0.1)

        self.class_names = None
        self.train_losses, self.train_accs, self.val_accs, self.val_losses = [], [], [], []

    def update_num_classes(self, num_classes):
        self.model.classifier = nn.Sequential(
            nn.LayerNorm(512),
            nn.Dropout(0.4),
            nn.Linear(512, 256),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_classes)
        ).to(device)
        print(f"🔄 Updated for {num_classes} classes")

    def prepare_data(self, paths, labels, class_names):
        self.image_paths = np.array(paths)
        self.labels = np.array(labels)
        self.class_names = class_names
        num_classes = len(np.unique(self.labels))
        self.update_num_classes(num_classes)
        return self.image_paths, self.labels

    def train(self, train_loader, val_loader):
        best_acc = 0
        best_state = None
        for epoch in range(Config.EPOCHS):
            self.model.train()
            total_loss, correct, total = 0, 0, 0

            pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{Config.EPOCHS}')
            for imgs, lbls in pbar:
                imgs, lbls = imgs.to(device), lbls.to(device)
                self.optimizer.zero_grad()

                if scaler is not None:
                    with torch.cuda.amp.autocast():
                        outs, _ = self.model(imgs)
                        loss = self.criterion(outs, lbls)
                    scaler.scale(loss).backward()
                    scaler.step(self.optimizer)
                    scaler.update()
                else:
                    outs, _ = self.model(imgs)
                    loss = self.criterion(outs, lbls)
                    loss.backward()
                    self.optimizer.step()

                total_loss += loss.item()
                pred = outs.argmax(1)
                correct += (pred == lbls).sum().item()
                total += lbls.size(0)

                pbar.set_postfix({'loss': f'{loss.item():.3f}',
                                  'acc': f'{correct/total:.3f}'})

            train_acc = correct / total
            avg_loss = total_loss / len(train_loader)
            val_acc, val_loss = self.validate(val_loader)

            self.scheduler.step()
            self.train_losses.append(avg_loss)
            self.train_accs.append(train_acc)
            self.val_accs.append(val_acc)
            self.val_losses.append(val_loss)

            if val_acc > best_acc:
                best_acc = val_acc
                best_state = self.model.state_dict()

            print(f'Epoch {epoch+1}: Train {train_acc:.4f} | Val {val_acc:.4f} | Loss {avg_loss:.4f}')

        if best_state is not None:
            self.model.load_state_dict(best_state)
        return best_acc

    @torch.no_grad()
    def validate(self, loader):
        self.model.eval()
        correct, total, loss_sum = 0, 0, 0
        for imgs, lbls in loader:
            imgs, lbls = imgs.to(device), lbls.to(device)
            outs, _ = self.model(imgs)
            loss = self.criterion(outs, lbls)
            loss_sum += loss.item()
            pred = outs.argmax(1)
            correct += (pred == lbls).sum().item()
            total += lbls.size(0)
        return correct / total, loss_sum / len(loader)

    @torch.no_grad()
    def test_full_metrics(self, loader):
        self.model.eval()
        all_preds, all_labels, all_probs = [], [], []
        for imgs, lbls in loader:
            imgs, lbls = imgs.to(device), lbls.to(device)
            outs, _ = self.model(imgs)
            probs = F.softmax(outs, dim=1)
            pred = outs.argmax(1)
            all_preds.extend(pred.cpu().numpy())
            all_labels.extend(lbls.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())

        return np.array(all_labels), np.array(all_preds), np.array(all_probs)

    @torch.no_grad()
    def predict_one_per_class(self, data_dir, output_dir="sample_predictions"):
        os.makedirs(output_dir, exist_ok=True)
        print("\n📸 Generating one prediction per class...")

        orig_paths, orig_labels, _ = detect_dataset_structure(data_dir)

        picked_paths, picked_labels = [], []
        seen_classes = set()
        for path, label in zip(orig_paths, orig_labels):
            if label not in seen_classes:
                picked_paths.append(path)
                picked_labels.append(label)
                seen_classes.add(label)
                if len(seen_classes) == len(self.class_names):
                    break

        _, val_test_transform = get_fast_transforms()
        pred_ds = WheatDiseaseDataset(picked_paths, np.array(picked_labels),
                                      val_test_transform)
        pred_loader = DataLoader(pred_ds, batch_size=1, shuffle=False)

        self.model.eval()

        results = []
        for i, (img, label) in enumerate(pred_loader):
            img = img.to(device)
            label = label.item()

            out, _ = self.model(img)
            prob = F.softmax(out, dim=1)
            pred = prob.argmax(1).item()
            confidence = prob.max().item()

            true_name = self.class_names[label]
            pred_name = self.class_names[pred]

            img_pil = Image.open(picked_paths[i]).convert('RGB').resize((224, 224))
            save_name = f"class_{label}_true-{true_name[:10]}_pred-{pred_name[:10]}_conf-{confidence:.1%}.png"
            save_path = os.path.join(output_dir, save_name)
            img_pil.save(save_path)

            results.append({
                "path": save_path,
                "true_label": label,
                "true_name": true_name,
                "pred_label": pred,
                "pred_name": pred_name,
                "confidence": confidence
            })

        print("\n📸 ONE PREDICTION PER CLASS (Original Dataset):")
        print("-" * 80)
        for r in results:
            status = "✅ CORRECT" if r['true_label'] == r['pred_label'] else "❌ WRONG"
            print(f"{status} | {r['true_name']:<12} → {r['pred_name']:<12} ({r['confidence']:.1%})")
            print(f"           📁 {os.path.basename(r['path'])}")
        print("-" * 80)

        return results

# ===================== HELPERS =====================

def split_data_single(paths, labels):
    train_val, test, train_val_l, test_l = train_test_split(
        paths, labels,
        test_size=Config.TEST_SPLIT_RATIO,
        stratify=labels,
        random_state=42
    )
    train, val, train_l, val_l = train_test_split(
        train_val, train_val_l,
        test_size=0.111,  # ~ 0.1 of original (since 0.9 * 0.111 ≈ 0.1)
        stratify=train_val_l,
        random_state=42
    )
    return train, val, test, train_l, val_l, test_l

def get_weighted_sampler(labels):
    class_counts = np.bincount(labels)
    class_weights = 1.0 / (class_counts + 1e-6)
    sample_weights = class_weights[labels]
    return WeightedRandomSampler(sample_weights, len(sample_weights))

# ===================== MAIN: 3-FOLD CV =====================

def train_fast_gan_3fold(data_dir="/kaggle/input/wheat-leaf-disease/Dataset"):
    print(f"🎯 Starting 3-FOLD GAN-ENHANCED training: {data_dir}")

    # Build GAN-augmented dataset once
    all_paths, all_labels, class_names = prepare_gan_augmented_data_simple(data_dir)
    all_paths = np.array(all_paths)
    all_labels = np.array(all_labels)

    skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
    fold_accuracies = []

    for fold_idx, (train_val_idx, test_idx) in enumerate(skf.split(all_paths, all_labels), start=1):
        print("\n" + "=" * 80)
        print(f"🧩 FOLD {fold_idx}/3")
        print("=" * 80)

        paths_train_val = all_paths[train_val_idx]
        labels_train_val = all_labels[train_val_idx]
        paths_test = all_paths[test_idx]
        labels_test = all_labels[test_idx]

        train_p, val_p, _, train_l, val_l, _ = split_data_single(
            paths_train_val, labels_train_val
        )
        test_p = paths_test
        test_l = labels_test

        print(f"📊 Fold {fold_idx} splits (Real + Synthetic):")
        print(f"   Train: {len(train_p)} | Val: {len(val_p)} | Test: {len(test_p)}")

        train_transform, val_test_transform = get_fast_transforms()
        train_ds = WheatDiseaseDataset(train_p, train_l, train_transform)
        val_ds = WheatDiseaseDataset(val_p, val_l, val_test_transform)
        test_ds = WheatDiseaseDataset(test_p, test_l, val_test_transform)

        sampler = get_weighted_sampler(train_l)
        train_loader = DataLoader(
            train_ds, Config.BATCH_SIZE, sampler=sampler,
            num_workers=Config.NUM_WORKERS, pin_memory=True
        )
        val_loader = DataLoader(
            val_ds, Config.BATCH_SIZE, shuffle=False,
            num_workers=Config.NUM_WORKERS, pin_memory=True
        )
        test_loader = DataLoader(
            test_ds, Config.BATCH_SIZE, shuffle=False,
            num_workers=Config.NUM_WORKERS, pin_memory=True
        )

        system = FastWheatClassifier(num_classes=len(class_names))
        system.prepare_data(all_paths, all_labels, class_names)

        # Reset optimizer & scheduler
        system.optimizer = optim.AdamW(system.model.parameters(),
                                       lr=Config.LEARNING_RATE,
                                       weight_decay=Config.WEIGHT_DECAY)
        system.scheduler = CosineAnnealingLR(system.optimizer, T_max=Config.EPOCHS)
        system.train_losses, system.train_accs, system.val_accs, system.val_losses = [], [], [], []

        print(f"🚀 CLASSIFIER TRAINING (GAN Data) - Fold {fold_idx}")
        best_val = system.train(train_loader, val_loader)

        print(f"\n📊 FULL EVALUATION + VISUALIZATION - Fold {fold_idx}")
        y_test, y_pred, y_proba = system.test_full_metrics(test_loader)

        visualizer = MetricsVisualizer(system.class_names, fold_idx=fold_idx)
        visualizer.plot_loss_curve(system.train_losses, system.val_losses)
        visualizer.plot_accuracy_curve(system.train_accs, system.val_accs)
        visualizer.plot_confusion_matrix(y_test, y_pred)
        visualizer.plot_precision_recall(y_test, y_pred)
        visualizer.plot_roc_curves(y_test, y_proba)
        visualizer.plot_metrics_summary(y_test, y_pred)
        visualizer.print_classification_report(y_test, y_pred)

        final_acc = accuracy_score(y_test, y_pred)
        fold_accuracies.append(final_acc)
        print(f"\n🎉 FOLD {fold_idx} TEST ACCURACY: {final_acc:.4f} ({final_acc:.1%})")

        # Optional: per-fold one-per-class predictions
        system.predict_one_per_class(
            data_dir, output_dir=f"sample_predictions_fold{fold_idx}"
        )

    print("\n" + "#" * 80)
    print("📦 3-FOLD RESULTS")
    for i, acc in enumerate(fold_accuracies, start=1):
        print(f"   Fold {i}: {acc:.4f} ({acc:.1%})")
    mean_acc = np.mean(fold_accuracies)
    std_acc = np.std(fold_accuracies)
    print(f"   Mean Acc: {mean_acc:.4f} ± {std_acc:.4f}")
    print("#" * 80 + "\n")

    return fold_accuracies, mean_acc, std_acc

if __name__ == "__main__":
    fold_accs, mean_acc, std_acc = train_fast_gan_3fold(
        "/kaggle/input/wheat-leaf-disease/Dataset"
    )
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
import numpy as np
from PIL import Image
import os
import random
import torchvision.transforms as transforms
import timm
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, \
    precision_recall_fscore_support, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import warnings
from torch.optim.lr_scheduler import CosineAnnealingLR
import albumentations as A
from albumentations.pytorch import ToTensorV2
from sklearn.preprocessing import label_binarize

warnings.filterwarnings("ignore")

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
scaler = torch.cuda.amp.GradScaler() if device.type == 'cuda' else None
print(f"Using device: {device}")

class Config:
    IMAGE_SIZE = 224
    BATCH_SIZE = 64
    NUM_WORKERS = 8
    NUM_CLASSES = 5

    # Reduced epochs for classifier
    EPOCHS = 15
    LEARNING_RATE = 3e-4
    WEIGHT_DECAY = 1e-4

    # Reduced GAN epochs
    GAN_EPOCHS = 20
    GAN_BATCH_SIZE = 32
    GAN_LR = 2e-4
    GAN_FEATURE_MATCHING = True
    GAN_AUGMENTATION_RATIO = 1

    TRAIN_SPLIT_RATIO = 0.8
    TEST_SPLIT_RATIO = 0.1
    VIT_BACKBONE = 'vit_base_patch16_224'
    CNN_BACKBONE = 'efficientnet_b2'
    LSTM_HIDDEN_SIZE = 256
    LSTM_NUM_LAYERS = 2
    LSTM_DROPOUT = 0.3
    SEQUENCE_LENGTH = 6

# ===================== GAN MODULES =====================

class ConditionalGenerator(nn.Module):
    def __init__(self, num_classes, latent_dim=100, img_channels=3):
        super().__init__()
        self.label_emb = nn.Embedding(num_classes, 64)

        self.model = nn.Sequential(
            nn.Linear(latent_dim + 64, 256 * 4 * 4),
            nn.ReLU(),
            nn.Unflatten(1, (256, 4, 4)),
            nn.ConvTranspose2d(256, 128, 4, 2, 1), nn.BatchNorm2d(128), nn.ReLU(),
            nn.ConvTranspose2d(128, 64, 4, 2, 1), nn.BatchNorm2d(64), nn.ReLU(),
            nn.ConvTranspose2d(64, 32, 4, 2, 1), nn.BatchNorm2d(32), nn.ReLU(),
            nn.ConvTranspose2d(32, img_channels, 4, 2, 1), nn.Tanh()
        )

    def forward(self, z, labels):
        c = self.label_emb(labels)
        x = torch.cat([z, c], dim=1)
        img = self.model(x)
        return img

class ConditionalDiscriminator(nn.Module):
    def __init__(self, num_classes, img_channels=3):
        super().__init__()
        self.label_emb = nn.Embedding(num_classes, 64)

        self.model = nn.Sequential(
            nn.Conv2d(img_channels + 64, 64, 4, 2, 1), nn.LeakyReLU(0.2),
            nn.Conv2d(64, 128, 4, 2, 1), nn.BatchNorm2d(128), nn.LeakyReLU(0.2),
            nn.Conv2d(128, 256, 4, 2, 1), nn.BatchNorm2d(256), nn.LeakyReLU(0.2),
            nn.Conv2d(256, 512, 4, 2, 1), nn.BatchNorm2d(512), nn.LeakyReLU(0.2),
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),
            nn.Linear(512, 1)
        )

    def forward(self, img, labels):
        c = self.label_emb(labels).view(labels.size(0), 64, 1, 1)
        c = c.repeat(1, 1, img.size(2), img.size(3))
        x = torch.cat([img, c], dim=1)
        return self.model(x)

class WheatDiseaseGAN:
    def __init__(self, num_classes):
        self.num_classes = num_classes
        self.generator = ConditionalGenerator(num_classes).to(device)
        self.discriminator = ConditionalDiscriminator(num_classes).to(device)

        self.g_optimizer = optim.Adam(self.generator.parameters(), lr=Config.GAN_LR, betas=(0.5, 0.999))
        self.d_optimizer = optim.Adam(self.discriminator.parameters(), lr=Config.GAN_LR, betas=(0.5, 0.999))

        print(f"🆕 GAN - G: {sum(p.numel() for p in self.generator.parameters())/1e6:.1f}M, "
              f"D: {sum(p.numel() for p in self.discriminator.parameters())/1e6:.1f}M")

    def train_gan(self, dataloader, epochs=Config.GAN_EPOCHS):
        print("🚀 Training Conditional GAN...")
        criterion = nn.BCEWithLogitsLoss()

        for epoch in range(epochs):
            g_loss_epoch, d_loss_epoch = 0, 0

            pbar = tqdm(dataloader, desc=f'GAN E{epoch+1}', leave=False)
            for real_imgs, labels in pbar:
                batch_size = real_imgs.size(0)
                real_imgs = real_imgs.to(device)
                labels = labels.to(device)

                real_labels = torch.ones(batch_size, 1).to(device)
                fake_labels = torch.zeros(batch_size, 1).to(device)

                # === TRAIN DISCRIMINATOR ===
                self.d_optimizer.zero_grad()
                real_pred = self.discriminator(real_imgs, labels)
                d_real_loss = criterion(real_pred, real_labels)

                z = torch.randn(batch_size, 100).to(device)
                fake_imgs = self.generator(z, labels)
                fake_pred = self.discriminator(fake_imgs.detach(), labels)
                d_fake_loss = criterion(fake_pred, fake_labels)

                d_loss = (d_real_loss + d_fake_loss) / 2
                d_loss.backward()
                self.d_optimizer.step()

                # === TRAIN GENERATOR ===
                self.g_optimizer.zero_grad()
                fake_pred = self.discriminator(fake_imgs, labels)
                g_loss = criterion(fake_pred, real_labels)

                if Config.GAN_FEATURE_MATCHING:
                    feat_real = torch.mean(real_imgs, dim=[2, 3], keepdim=True)
                    feat_fake = torch.mean(fake_imgs, dim=[2, 3], keepdim=True)
                    fm_loss = F.mse_loss(feat_fake, feat_real)
                    g_loss += 0.1 * fm_loss

                g_loss.backward()
                self.g_optimizer.step()

                g_loss_epoch += g_loss.item()
                d_loss_epoch += d_loss.item()
                pbar.set_postfix({'D': f'{d_loss.item():.3f}', 'G': f'{g_loss.item():.3f}'})

            if epoch % 5 == 0 or epoch == epochs - 1:
                print(f'GAN E[{epoch+1}/{epochs}] D:{d_loss_epoch/len(dataloader):.3f} '
                      f'G:{g_loss_epoch/len(dataloader):.3f}')

        print("✅ GAN trained!")

    @torch.no_grad()
    def generate_synthetic_data(self, num_samples_per_class):
        print("🎨 Generating synthetic images...")
        synthetic_paths, synthetic_labels = [], []
        os.makedirs("synthetic_data", exist_ok=True)

        self.generator.eval()
        with torch.no_grad():
            for class_idx in range(self.num_classes):
                labels = torch.full((num_samples_per_class,), class_idx,
                                    dtype=torch.long).to(device)
                z = torch.randn(num_samples_per_class, 100).to(device)
                fake_imgs = self.generator(z, labels)

                class_dir = f"synthetic_data/class_{class_idx}"
                os.makedirs(class_dir, exist_ok=True)

                for i, img in enumerate(fake_imgs):
                    img = (img * 0.5 + 0.5).clamp(0, 1)
                    # keep all channels
                    img_pil = transforms.ToPILImage()(img.cpu())
                    path = os.path.join(class_dir, f"syn_{i:04d}.png")
                    img_pil.save(path)
                    synthetic_paths.append(path)
                    synthetic_labels.append(class_idx)

        print(f"✅ Generated {len(synthetic_paths)} images")
        return synthetic_paths, np.array(synthetic_labels)

# ===================== DATA & TRANSFORMS =====================

def detect_dataset_structure(data_dir):
    class_names, image_paths, labels = [], [], []
    print(f"🔍 Scanning: {data_dir}")

    for root, dirs, files in os.walk(data_dir):
        image_files = [f for f in files if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        if not image_files:
            continue

        class_name = os.path.basename(root)
        if class_name not in class_names:
            class_names.append(class_name)
            class_names.sort()

        class_idx = class_names.index(class_name)
        for img_file in image_files:
            image_paths.append(os.path.join(root, img_file))
            labels.append(class_idx)

    print(f"✅ Found {len(class_names)} classes: {class_names}")
    print(f"📊 Total images: {len(image_paths)}")
    return image_paths, np.array(labels), class_names

def get_fast_transforms():
    train_transform = A.Compose([
        A.Resize(Config.IMAGE_SIZE, Config.IMAGE_SIZE),
        A.HorizontalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.RandomBrightnessContrast(p=0.3),
        A.Normalize(mean=[0.485, 0.456, 0.406],
                    std=[0.229, 0.224, 0.225]),
        ToTensorV2()
    ])
    val_test_transform = A.Compose([
        A.Resize(Config.IMAGE_SIZE, Config.IMAGE_SIZE),
        A.Normalize(mean=[0.485, 0.456, 0.406],
                    std=[0.229, 0.224, 0.225]),
        ToTensorV2()
    ])
    return train_transform, val_test_transform

class WheatDiseaseDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = list(image_paths)
        self.labels = np.array(labels)
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        try:
            image = Image.open(self.image_paths[idx]).convert('RGB')
            image = np.array(image)
            label = int(self.labels[idx])
            if self.transform:
                augmented = self.transform(image=image)
                image = augmented['image']
            return image, label
        except:
            dummy = np.full((Config.IMAGE_SIZE, Config.IMAGE_SIZE, 3),
                            128, dtype=np.uint8)
            return torch.tensor(dummy.transpose(2, 0, 1) / 255.0), 0

def prepare_gan_augmented_data_simple(data_dir):
    orig_paths, orig_labels, class_names = detect_dataset_structure(data_dir)
    if len(orig_paths) == 0:
        raise ValueError("No images found!")

    _, val_test_transform = get_fast_transforms()

    gan_dataset = WheatDiseaseDataset(orig_paths, orig_labels, val_test_transform)
    gan_loader = DataLoader(
        gan_dataset,
        batch_size=Config.GAN_BATCH_SIZE,
        shuffle=True,
        num_workers=4,
        pin_memory=True
    )

    gan_trainer = WheatDiseaseGAN(len(class_names))
    gan_trainer.train_gan(gan_loader, epochs=Config.GAN_EPOCHS)

    orig_counts = np.bincount(orig_labels)
    target_per_class = min(int(np.min(orig_counts) * Config.GAN_AUGMENTATION_RATIO), 400)
    syn_paths, syn_labels = gan_trainer.generate_synthetic_data(target_per_class)

    all_paths = orig_paths + syn_paths
    all_labels = np.concatenate([orig_labels, syn_labels])

    print(f"📊 GAN Augmented: {len(orig_paths)} → {len(all_paths)} images")
    return all_paths, all_labels, class_names

# ===================== MODEL (CNN + ViT + LSTM) =====================

class CNNFeatureExtractor(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = timm.create_model(Config.CNN_BACKBONE, pretrained=True, num_classes=0)
        for param in self.backbone.parameters():
            param.requires_grad = False
        # Unfreeze last 2 blocks if exist
        if hasattr(self.backbone, "blocks"):
            for param in self.backbone.blocks[-2:].parameters():
                param.requires_grad = True

        dummy = torch.randn(1, 3, Config.IMAGE_SIZE, Config.IMAGE_SIZE)
        with torch.no_grad():
            feats = self.backbone.forward_features(dummy)
        self.feature_dim = feats.shape[1] if feats.ndim == 4 else feats.shape[-1]
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.proj = nn.Sequential(
            nn.Linear(self.feature_dim, 512),
            nn.GELU(),
            nn.Dropout(0.3)
        )

    def forward(self, x):
        feats = self.backbone.forward_features(x)
        if feats.ndim == 4:
            pooled = self.pool(feats).flatten(1)
        else:
            pooled = feats
        return self.proj(pooled)

class ViTFeatureExtractor(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = timm.create_model(Config.VIT_BACKBONE, pretrained=True, num_classes=0)
        for param in self.backbone.parameters():
            param.requires_grad = False
        if hasattr(self.backbone, "blocks"):
            for param in self.backbone.blocks[-2:].parameters():
                param.requires_grad = True

    def forward(self, x):
        return self.backbone(x)

class FastAttention(nn.Module):
    def __init__(self, d_model, num_heads=8):
        super().__init__()
        self.attn = nn.MultiheadAttention(d_model, num_heads,
                                          batch_first=True, dropout=0.1)

    def forward(self, x):
        attn_out, attn_weights = self.attn(x, x, x)
        return attn_out[:, -1:], attn_weights.mean(dim=1)

class FastHybridClassifier(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.cnn = CNNFeatureExtractor()
        self.vit = ViTFeatureExtractor()
        self.lstm = nn.LSTM(512 + 768, 256, 2,
                            batch_first=True, bidirectional=True, dropout=0.2)
        self.attention = FastAttention(512)
        self.classifier = nn.Sequential(
            nn.LayerNorm(512),
            nn.Dropout(0.4),
            nn.Linear(512, 256),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        cnn_feat = self.cnn(x)
        vit_feat = self.vit(x)
        fused = torch.cat([vit_feat, cnn_feat], dim=1).unsqueeze(1)
        seq = [fused + torch.randn_like(fused) * 0.02
               for _ in range(Config.SEQUENCE_LENGTH)]
        seq = torch.cat(seq, dim=1)
        lstm_out, _ = self.lstm(seq)
        attn_out, attn_w = self.attention(lstm_out)
        return self.classifier(attn_out.squeeze(1)), attn_w

# ===================== METRICS & VISUALIZATION =====================

class MetricsVisualizer:
    def __init__(self, class_names, fold_idx=None):
        self.class_names = class_names
        self.fold_idx = fold_idx

    def _prefix(self, name):
        if self.fold_idx is None:
            return name
        base, ext = os.path.splitext(name)
        return f"fold{self.fold_idx}_{base}{ext}"

    def plot_loss_curve(self, train_losses, val_losses):
        plt.figure(figsize=(10, 6))
        plt.plot(train_losses, label='Train Loss', linewidth=3, color='blue')
        plt.plot(val_losses, label='Val Loss', linewidth=3, color='red')
        plt.title('🗺️ Training & Validation Loss', fontsize=16, fontweight='bold', pad=20)
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        fname = self._prefix('1_loss_curve.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_accuracy_curve(self, train_accs, val_accs):
        plt.figure(figsize=(10, 6))
        plt.plot(train_accs, label='Train Acc', linewidth=3, color='green')
        plt.plot(val_accs, label='Val Acc', linewidth=3, color='orange')
        plt.title('📈 Training & Validation Accuracy', fontsize=16, fontweight='bold', pad=20)
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        fname = self._prefix('2_accuracy_curve.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_confusion_matrix(self, y_test, y_pred):
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=self.class_names, yticklabels=self.class_names)
        plt.title('🔍 Confusion Matrix (Test Set)', fontsize=16, fontweight='bold')
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.tight_layout()
        fname = self._prefix('3_confusion_matrix.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_precision_recall(self, y_test, y_pred):
        precision, recall, _, _ = precision_recall_fscore_support(
            y_test, y_pred, average=None)
        x_pos = np.arange(len(self.class_names))
        width = 0.35

        plt.figure(figsize=(12, 6))
        plt.bar(x_pos - width / 2, precision, width,
                label='Precision', alpha=0.8, color='skyblue')
        plt.bar(x_pos + width / 2, recall, width,
                label='Recall', alpha=0.8, color='salmon')
        plt.title('⚖️ Precision & Recall per Class', fontsize=16, fontweight='bold')
        plt.xlabel('Classes')
        plt.ylabel('Score')
        plt.xticks(x_pos, [name[:12] for name in self.class_names], rotation=45)
        plt.legend()
        plt.ylim(0, 1)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        fname = self._prefix('4_precision_recall.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_roc_curves(self, y_test, y_proba):
        plt.figure(figsize=(12, 10))
        y_bin = label_binarize(y_test, classes=range(len(self.class_names)))
        colors = plt.cm.Set1(np.linspace(0, 1, len(self.class_names)))

        for i, color in enumerate(colors):
            if y_bin[:, i].sum() > 0:
                fpr, tpr, _ = roc_curve(y_bin[:, i], y_proba[:, i])
                roc_auc = auc(fpr, tpr)
                plt.plot(fpr, tpr, color=color, lw=2.5,
                         label=f'{self.class_names[i][:15]} (AUC = {roc_auc:.3f})')

        plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Random Classifier')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate', fontsize=12)
        plt.ylabel('True Positive Rate', fontsize=12)
        plt.title('🎯 Multi-Class ROC Curves (GAN-Augmented Test Set)',
                  fontsize=16, fontweight='bold', pad=20)
        plt.legend(loc="lower right", fontsize=11)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        fname = self._prefix('6_roc_curves.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_metrics_summary(self, y_test, y_pred):
        precision_macro, recall_macro, f1_macro, _ = precision_recall_fscore_support(
            y_test, y_pred, average='macro')
        precision_weighted, recall_weighted, f1_weighted, _ = precision_recall_fscore_support(
            y_test, y_pred, average='weighted')
        final_acc = accuracy_score(y_test, y_pred)

        metrics_data = [
            ['Metric', 'Macro Avg', 'Weighted Avg'],
            ['Accuracy', f'{final_acc:.3f}', f'{final_acc:.3f}'],
            ['Precision', f'{precision_macro:.3f}', f'{precision_weighted:.3f}'],
            ['Recall', f'{recall_macro:.3f}', f'{recall_weighted:.3f}'],
            ['F1-Score', f'{f1_macro:.3f}', f'{f1_weighted:.3f}']
        ]

        plt.figure(figsize=(10, 6))
        table = plt.table(cellText=metrics_data[1:],
                          colLabels=metrics_data[0],
                          cellLoc='center', loc='center')
        table.auto_set_font_size(False)
        table.set_fontsize(12)
        table.scale(1, 2.5)
        plt.title('📊 GAN-Augmented Metrics Summary', fontsize=16,
                  fontweight='bold', pad=20)
        plt.axis('off')
        plt.tight_layout()
        fname = self._prefix('7_metrics_summary.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def print_classification_report(self, y_test, y_pred):
        report = classification_report(y_test, y_pred,
                                       target_names=self.class_names,
                                       digits=4)
        print("\n" + "=" * 80)
        print("📋 GAN-AUGMENTED CLASSIFICATION REPORT")
        print("=" * 80)
        print(report)
        fname = self._prefix('classification_report.txt')
        with open(fname, 'w') as f:
            f.write(report)
        print(f"✅ Saved: {fname}")
        print("=" * 80)

# ===================== CLASSIFIER WRAPPER =====================

class FastWheatClassifier:
    def __init__(self, num_classes=5):
        self.model = FastHybridClassifier(num_classes).to(device)
        print(f"Classifier: {sum(p.numel() for p in self.model.parameters())/1e6:.1f}M params")

        self.optimizer = optim.AdamW(self.model.parameters(),
                                     lr=Config.LEARNING_RATE,
                                     weight_decay=Config.WEIGHT_DECAY)
        self.scheduler = CosineAnnealingLR(self.optimizer, T_max=Config.EPOCHS)
        self.criterion = nn.CrossEntropyLoss(label_smoothing=0.1)

        self.class_names = None
        self.train_losses, self.train_accs, self.val_accs, self.val_losses = [], [], [], []

    def update_num_classes(self, num_classes):
        self.model.classifier = nn.Sequential(
            nn.LayerNorm(512),
            nn.Dropout(0.4),
            nn.Linear(512, 256),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_classes)
        ).to(device)
        print(f"🔄 Updated for {num_classes} classes")

    def prepare_data(self, paths, labels, class_names):
        self.image_paths = np.array(paths)
        self.labels = np.array(labels)
        self.class_names = class_names
        num_classes = len(np.unique(self.labels))
        self.update_num_classes(num_classes)
        return self.image_paths, self.labels

    def train(self, train_loader, val_loader):
        best_acc = 0
        best_state = None
        for epoch in range(Config.EPOCHS):
            self.model.train()
            total_loss, correct, total = 0, 0, 0

            pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{Config.EPOCHS}')
            for imgs, lbls in pbar:
                imgs, lbls = imgs.to(device), lbls.to(device)
                self.optimizer.zero_grad()

                if scaler is not None:
                    with torch.cuda.amp.autocast():
                        outs, _ = self.model(imgs)
                        loss = self.criterion(outs, lbls)
                    scaler.scale(loss).backward()
                    scaler.step(self.optimizer)
                    scaler.update()
                else:
                    outs, _ = self.model(imgs)
                    loss = self.criterion(outs, lbls)
                    loss.backward()
                    self.optimizer.step()

                total_loss += loss.item()
                pred = outs.argmax(1)
                correct += (pred == lbls).sum().item()
                total += lbls.size(0)

                pbar.set_postfix({'loss': f'{loss.item():.3f}',
                                  'acc': f'{correct/total:.3f}'})

            train_acc = correct / total
            avg_loss = total_loss / len(train_loader)
            val_acc, val_loss = self.validate(val_loader)

            self.scheduler.step()
            self.train_losses.append(avg_loss)
            self.train_accs.append(train_acc)
            self.val_accs.append(val_acc)
            self.val_losses.append(val_loss)

            if val_acc > best_acc:
                best_acc = val_acc
                best_state = self.model.state_dict()

            print(f'Epoch {epoch+1}: Train {train_acc:.4f} | Val {val_acc:.4f} | Loss {avg_loss:.4f}')

        if best_state is not None:
            self.model.load_state_dict(best_state)
        return best_acc

    @torch.no_grad()
    def validate(self, loader):
        self.model.eval()
        correct, total, loss_sum = 0, 0, 0
        for imgs, lbls in loader:
            imgs, lbls = imgs.to(device), lbls.to(device)
            outs, _ = self.model(imgs)
            loss = self.criterion(outs, lbls)
            loss_sum += loss.item()
            pred = outs.argmax(1)
            correct += (pred == lbls).sum().item()
            total += lbls.size(0)
        return correct / total, loss_sum / len(loader)

    @torch.no_grad()
    def test_full_metrics(self, loader):
        self.model.eval()
        all_preds, all_labels, all_probs = [], [], []
        for imgs, lbls in loader:
            imgs, lbls = imgs.to(device), lbls.to(device)
            outs, _ = self.model(imgs)
            probs = F.softmax(outs, dim=1)
            pred = outs.argmax(1)
            all_preds.extend(pred.cpu().numpy())
            all_labels.extend(lbls.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())

        return np.array(all_labels), np.array(all_preds), np.array(all_probs)

    @torch.no_grad()
    def predict_one_per_class(self, data_dir, output_dir="sample_predictions"):
        os.makedirs(output_dir, exist_ok=True)
        print("\n📸 Generating one prediction per class...")

        orig_paths, orig_labels, _ = detect_dataset_structure(data_dir)

        picked_paths, picked_labels = [], []
        seen_classes = set()
        for path, label in zip(orig_paths, orig_labels):
            if label not in seen_classes:
                picked_paths.append(path)
                picked_labels.append(label)
                seen_classes.add(label)
                if len(seen_classes) == len(self.class_names):
                    break

        _, val_test_transform = get_fast_transforms()
        pred_ds = WheatDiseaseDataset(picked_paths, np.array(picked_labels),
                                      val_test_transform)
        pred_loader = DataLoader(pred_ds, batch_size=1, shuffle=False)

        self.model.eval()

        results = []
        for i, (img, label) in enumerate(pred_loader):
            img = img.to(device)
            label = label.item()

            out, _ = self.model(img)
            prob = F.softmax(out, dim=1)
            pred = prob.argmax(1).item()
            confidence = prob.max().item()

            true_name = self.class_names[label]
            pred_name = self.class_names[pred]

            img_pil = Image.open(picked_paths[i]).convert('RGB').resize((224, 224))
            save_name = f"class_{label}_true-{true_name[:10]}_pred-{pred_name[:10]}_conf-{confidence:.1%}.png"
            save_path = os.path.join(output_dir, save_name)
            img_pil.save(save_path)

            results.append({
                "path": save_path,
                "true_label": label,
                "true_name": true_name,
                "pred_label": pred,
                "pred_name": pred_name,
                "confidence": confidence
            })

        print("\n📸 ONE PREDICTION PER CLASS (Original Dataset):")
        print("-" * 80)
        for r in results:
            status = "✅ CORRECT" if r['true_label'] == r['pred_label'] else "❌ WRONG"
            print(f"{status} | {r['true_name']:<12} → {r['pred_name']:<12} ({r['confidence']:.1%})")
            print(f"           📁 {os.path.basename(r['path'])}")
        print("-" * 80)

        return results

# ===================== HELPERS =====================

def split_data_single(paths, labels):
    train_val, test, train_val_l, test_l = train_test_split(
        paths, labels,
        test_size=Config.TEST_SPLIT_RATIO,
        stratify=labels,
        random_state=42
    )
    train, val, train_l, val_l = train_test_split(
        train_val, train_val_l,
        test_size=0.111,  # ~ 0.1 of original (since 0.9 * 0.111 ≈ 0.1)
        stratify=train_val_l,
        random_state=42
    )
    return train, val, test, train_l, val_l, test_l

def get_weighted_sampler(labels):
    class_counts = np.bincount(labels)
    class_weights = 1.0 / (class_counts + 1e-6)
    sample_weights = class_weights[labels]
    return WeightedRandomSampler(sample_weights, len(sample_weights))

# ===================== MAIN: 3-FOLD CV =====================

def train_fast_gan_3fold(data_dir="/kaggle/input/large-wheat-disease-classification-dataset/Large Wheat Disease Classification Dataset"):
    print(f"🎯 Starting 3-FOLD GAN-ENHANCED training: {data_dir}")

    # Build GAN-augmented dataset once
    all_paths, all_labels, class_names = prepare_gan_augmented_data_simple(data_dir)
    all_paths = np.array(all_paths)
    all_labels = np.array(all_labels)

    skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
    fold_accuracies = []

    for fold_idx, (train_val_idx, test_idx) in enumerate(skf.split(all_paths, all_labels), start=1):
        print("\n" + "=" * 80)
        print(f"🧩 FOLD {fold_idx}/3")
        print("=" * 80)

        paths_train_val = all_paths[train_val_idx]
        labels_train_val = all_labels[train_val_idx]
        paths_test = all_paths[test_idx]
        labels_test = all_labels[test_idx]

        train_p, val_p, _, train_l, val_l, _ = split_data_single(
            paths_train_val, labels_train_val
        )
        test_p = paths_test
        test_l = labels_test

        print(f"📊 Fold {fold_idx} splits (Real + Synthetic):")
        print(f"   Train: {len(train_p)} | Val: {len(val_p)} | Test: {len(test_p)}")

        train_transform, val_test_transform = get_fast_transforms()
        train_ds = WheatDiseaseDataset(train_p, train_l, train_transform)
        val_ds = WheatDiseaseDataset(val_p, val_l, val_test_transform)
        test_ds = WheatDiseaseDataset(test_p, test_l, val_test_transform)

        sampler = get_weighted_sampler(train_l)
        train_loader = DataLoader(
            train_ds, Config.BATCH_SIZE, sampler=sampler,
            num_workers=Config.NUM_WORKERS, pin_memory=True
        )
        val_loader = DataLoader(
            val_ds, Config.BATCH_SIZE, shuffle=False,
            num_workers=Config.NUM_WORKERS, pin_memory=True
        )
        test_loader = DataLoader(
            test_ds, Config.BATCH_SIZE, shuffle=False,
            num_workers=Config.NUM_WORKERS, pin_memory=True
        )

        system = FastWheatClassifier(num_classes=len(class_names))
        system.prepare_data(all_paths, all_labels, class_names)

        # Reset optimizer & scheduler
        system.optimizer = optim.AdamW(system.model.parameters(),
                                       lr=Config.LEARNING_RATE,
                                       weight_decay=Config.WEIGHT_DECAY)
        system.scheduler = CosineAnnealingLR(system.optimizer, T_max=Config.EPOCHS)
        system.train_losses, system.train_accs, system.val_accs, system.val_losses = [], [], [], []

        print(f"🚀 CLASSIFIER TRAINING (GAN Data) - Fold {fold_idx}")
        best_val = system.train(train_loader, val_loader)

        print(f"\n📊 FULL EVALUATION + VISUALIZATION - Fold {fold_idx}")
        y_test, y_pred, y_proba = system.test_full_metrics(test_loader)

        visualizer = MetricsVisualizer(system.class_names, fold_idx=fold_idx)
        visualizer.plot_loss_curve(system.train_losses, system.val_losses)
        visualizer.plot_accuracy_curve(system.train_accs, system.val_accs)
        visualizer.plot_confusion_matrix(y_test, y_pred)
        visualizer.plot_precision_recall(y_test, y_pred)
        visualizer.plot_roc_curves(y_test, y_proba)
        visualizer.plot_metrics_summary(y_test, y_pred)
        visualizer.print_classification_report(y_test, y_pred)

        final_acc = accuracy_score(y_test, y_pred)
        fold_accuracies.append(final_acc)
        print(f"\n🎉 FOLD {fold_idx} TEST ACCURACY: {final_acc:.4f} ({final_acc:.1%})")

        # Optional: per-fold one-per-class predictions
        system.predict_one_per_class(
            data_dir, output_dir=f"sample_predictions_fold{fold_idx}"
        )

    print("\n" + "#" * 80)
    print("📦 3-FOLD RESULTS")
    for i, acc in enumerate(fold_accuracies, start=1):
        print(f"   Fold {i}: {acc:.4f} ({acc:.1%})")
    mean_acc = np.mean(fold_accuracies)
    std_acc = np.std(fold_accuracies)
    print(f"   Mean Acc: {mean_acc:.4f} ± {std_acc:.4f}")
    print("#" * 80 + "\n")

    return fold_accuracies, mean_acc, std_acc

if __name__ == "__main__":
    fold_accs, mean_acc, std_acc = train_fast_gan_3fold(
        "/kaggle/input/large-wheat-disease-classification-dataset/Large Wheat Disease Classification Dataset"
    )
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
import numpy as np
from PIL import Image
import os
import random
import torchvision.transforms as transforms
import timm
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, \
    precision_recall_fscore_support, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import warnings
from torch.optim.lr_scheduler import CosineAnnealingLR
import albumentations as A
from albumentations.pytorch import ToTensorV2
from sklearn.preprocessing import label_binarize

warnings.filterwarnings("ignore")

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
scaler = torch.cuda.amp.GradScaler() if device.type == 'cuda' else None
print(f"Using device: {device}")

class Config:
    IMAGE_SIZE = 224
    BATCH_SIZE = 64
    NUM_WORKERS = 8
    NUM_CLASSES = 5

    # Reduced epochs for classifier
    EPOCHS = 15
    LEARNING_RATE = 3e-4
    WEIGHT_DECAY = 1e-4

    # Reduced GAN epochs
    GAN_EPOCHS = 20
    GAN_BATCH_SIZE = 32
    GAN_LR = 2e-4
    GAN_FEATURE_MATCHING = True
    GAN_AUGMENTATION_RATIO = 1

    TRAIN_SPLIT_RATIO = 0.8
    TEST_SPLIT_RATIO = 0.1
    VIT_BACKBONE = 'vit_base_patch16_224'
    CNN_BACKBONE = 'efficientnet_b2'
    LSTM_HIDDEN_SIZE = 256
    LSTM_NUM_LAYERS = 2
    LSTM_DROPOUT = 0.3
    SEQUENCE_LENGTH = 6

# ===================== GAN MODULES =====================

class ConditionalGenerator(nn.Module):
    def __init__(self, num_classes, latent_dim=100, img_channels=3):
        super().__init__()
        self.label_emb = nn.Embedding(num_classes, 64)

        self.model = nn.Sequential(
            nn.Linear(latent_dim + 64, 256 * 4 * 4),
            nn.ReLU(),
            nn.Unflatten(1, (256, 4, 4)),
            nn.ConvTranspose2d(256, 128, 4, 2, 1), nn.BatchNorm2d(128), nn.ReLU(),
            nn.ConvTranspose2d(128, 64, 4, 2, 1), nn.BatchNorm2d(64), nn.ReLU(),
            nn.ConvTranspose2d(64, 32, 4, 2, 1), nn.BatchNorm2d(32), nn.ReLU(),
            nn.ConvTranspose2d(32, img_channels, 4, 2, 1), nn.Tanh()
        )

    def forward(self, z, labels):
        c = self.label_emb(labels)
        x = torch.cat([z, c], dim=1)
        img = self.model(x)
        return img

class ConditionalDiscriminator(nn.Module):
    def __init__(self, num_classes, img_channels=3):
        super().__init__()
        self.label_emb = nn.Embedding(num_classes, 64)

        self.model = nn.Sequential(
            nn.Conv2d(img_channels + 64, 64, 4, 2, 1), nn.LeakyReLU(0.2),
            nn.Conv2d(64, 128, 4, 2, 1), nn.BatchNorm2d(128), nn.LeakyReLU(0.2),
            nn.Conv2d(128, 256, 4, 2, 1), nn.BatchNorm2d(256), nn.LeakyReLU(0.2),
            nn.Conv2d(256, 512, 4, 2, 1), nn.BatchNorm2d(512), nn.LeakyReLU(0.2),
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),
            nn.Linear(512, 1)
        )

    def forward(self, img, labels):
        c = self.label_emb(labels).view(labels.size(0), 64, 1, 1)
        c = c.repeat(1, 1, img.size(2), img.size(3))
        x = torch.cat([img, c], dim=1)
        return self.model(x)

class WheatDiseaseGAN:
    def __init__(self, num_classes):
        self.num_classes = num_classes
        self.generator = ConditionalGenerator(num_classes).to(device)
        self.discriminator = ConditionalDiscriminator(num_classes).to(device)

        self.g_optimizer = optim.Adam(self.generator.parameters(), lr=Config.GAN_LR, betas=(0.5, 0.999))
        self.d_optimizer = optim.Adam(self.discriminator.parameters(), lr=Config.GAN_LR, betas=(0.5, 0.999))

        print(f"🆕 GAN - G: {sum(p.numel() for p in self.generator.parameters())/1e6:.1f}M, "
              f"D: {sum(p.numel() for p in self.discriminator.parameters())/1e6:.1f}M")

    def train_gan(self, dataloader, epochs=Config.GAN_EPOCHS):
        print("🚀 Training Conditional GAN...")
        criterion = nn.BCEWithLogitsLoss()

        for epoch in range(epochs):
            g_loss_epoch, d_loss_epoch = 0, 0

            pbar = tqdm(dataloader, desc=f'GAN E{epoch+1}', leave=False)
            for real_imgs, labels in pbar:
                batch_size = real_imgs.size(0)
                real_imgs = real_imgs.to(device)
                labels = labels.to(device)

                real_labels = torch.ones(batch_size, 1).to(device)
                fake_labels = torch.zeros(batch_size, 1).to(device)

                # === TRAIN DISCRIMINATOR ===
                self.d_optimizer.zero_grad()
                real_pred = self.discriminator(real_imgs, labels)
                d_real_loss = criterion(real_pred, real_labels)

                z = torch.randn(batch_size, 100).to(device)
                fake_imgs = self.generator(z, labels)
                fake_pred = self.discriminator(fake_imgs.detach(), labels)
                d_fake_loss = criterion(fake_pred, fake_labels)

                d_loss = (d_real_loss + d_fake_loss) / 2
                d_loss.backward()
                self.d_optimizer.step()

                # === TRAIN GENERATOR ===
                self.g_optimizer.zero_grad()
                fake_pred = self.discriminator(fake_imgs, labels)
                g_loss = criterion(fake_pred, real_labels)

                if Config.GAN_FEATURE_MATCHING:
                    feat_real = torch.mean(real_imgs, dim=[2, 3], keepdim=True)
                    feat_fake = torch.mean(fake_imgs, dim=[2, 3], keepdim=True)
                    fm_loss = F.mse_loss(feat_fake, feat_real)
                    g_loss += 0.1 * fm_loss

                g_loss.backward()
                self.g_optimizer.step()

                g_loss_epoch += g_loss.item()
                d_loss_epoch += d_loss.item()
                pbar.set_postfix({'D': f'{d_loss.item():.3f}', 'G': f'{g_loss.item():.3f}'})

            if epoch % 5 == 0 or epoch == epochs - 1:
                print(f'GAN E[{epoch+1}/{epochs}] D:{d_loss_epoch/len(dataloader):.3f} '
                      f'G:{g_loss_epoch/len(dataloader):.3f}')

        print("✅ GAN trained!")

    @torch.no_grad()
    def generate_synthetic_data(self, num_samples_per_class):
        print("🎨 Generating synthetic images...")
        synthetic_paths, synthetic_labels = [], []
        os.makedirs("synthetic_data", exist_ok=True)

        self.generator.eval()
        with torch.no_grad():
            for class_idx in range(self.num_classes):
                labels = torch.full((num_samples_per_class,), class_idx,
                                    dtype=torch.long).to(device)
                z = torch.randn(num_samples_per_class, 100).to(device)
                fake_imgs = self.generator(z, labels)

                class_dir = f"synthetic_data/class_{class_idx}"
                os.makedirs(class_dir, exist_ok=True)

                for i, img in enumerate(fake_imgs):
                    img = (img * 0.5 + 0.5).clamp(0, 1)
                    # keep all channels
                    img_pil = transforms.ToPILImage()(img.cpu())
                    path = os.path.join(class_dir, f"syn_{i:04d}.png")
                    img_pil.save(path)
                    synthetic_paths.append(path)
                    synthetic_labels.append(class_idx)

        print(f"✅ Generated {len(synthetic_paths)} images")
        return synthetic_paths, np.array(synthetic_labels)

# ===================== DATA & TRANSFORMS =====================

def detect_dataset_structure(data_dir):
    class_names, image_paths, labels = [], [], []
    print(f"🔍 Scanning: {data_dir}")

    for root, dirs, files in os.walk(data_dir):
        image_files = [f for f in files if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        if not image_files:
            continue

        class_name = os.path.basename(root)
        if class_name not in class_names:
            class_names.append(class_name)
            class_names.sort()

        class_idx = class_names.index(class_name)
        for img_file in image_files:
            image_paths.append(os.path.join(root, img_file))
            labels.append(class_idx)

    print(f"✅ Found {len(class_names)} classes: {class_names}")
    print(f"📊 Total images: {len(image_paths)}")
    return image_paths, np.array(labels), class_names

def get_fast_transforms():
    train_transform = A.Compose([
        A.Resize(Config.IMAGE_SIZE, Config.IMAGE_SIZE),
        A.HorizontalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.RandomBrightnessContrast(p=0.3),
        A.Normalize(mean=[0.485, 0.456, 0.406],
                    std=[0.229, 0.224, 0.225]),
        ToTensorV2()
    ])
    val_test_transform = A.Compose([
        A.Resize(Config.IMAGE_SIZE, Config.IMAGE_SIZE),
        A.Normalize(mean=[0.485, 0.456, 0.406],
                    std=[0.229, 0.224, 0.225]),
        ToTensorV2()
    ])
    return train_transform, val_test_transform

class WheatDiseaseDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = list(image_paths)
        self.labels = np.array(labels)
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        try:
            image = Image.open(self.image_paths[idx]).convert('RGB')
            image = np.array(image)
            label = int(self.labels[idx])
            if self.transform:
                augmented = self.transform(image=image)
                image = augmented['image']
            return image, label
        except:
            dummy = np.full((Config.IMAGE_SIZE, Config.IMAGE_SIZE, 3),
                            128, dtype=np.uint8)
            return torch.tensor(dummy.transpose(2, 0, 1) / 255.0), 0

def prepare_gan_augmented_data_simple(data_dir):
    orig_paths, orig_labels, class_names = detect_dataset_structure(data_dir)
    if len(orig_paths) == 0:
        raise ValueError("No images found!")

    _, val_test_transform = get_fast_transforms()

    gan_dataset = WheatDiseaseDataset(orig_paths, orig_labels, val_test_transform)
    gan_loader = DataLoader(
        gan_dataset,
        batch_size=Config.GAN_BATCH_SIZE,
        shuffle=True,
        num_workers=4,
        pin_memory=True
    )

    gan_trainer = WheatDiseaseGAN(len(class_names))
    gan_trainer.train_gan(gan_loader, epochs=Config.GAN_EPOCHS)

    orig_counts = np.bincount(orig_labels)
    target_per_class = min(int(np.min(orig_counts) * Config.GAN_AUGMENTATION_RATIO), 400)
    syn_paths, syn_labels = gan_trainer.generate_synthetic_data(target_per_class)

    all_paths = orig_paths + syn_paths
    all_labels = np.concatenate([orig_labels, syn_labels])

    print(f"📊 GAN Augmented: {len(orig_paths)} → {len(all_paths)} images")
    return all_paths, all_labels, class_names

# ===================== MODEL (CNN + ViT + LSTM) =====================

class CNNFeatureExtractor(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = timm.create_model(Config.CNN_BACKBONE, pretrained=True, num_classes=0)
        for param in self.backbone.parameters():
            param.requires_grad = False
        # Unfreeze last 2 blocks if exist
        if hasattr(self.backbone, "blocks"):
            for param in self.backbone.blocks[-2:].parameters():
                param.requires_grad = True

        dummy = torch.randn(1, 3, Config.IMAGE_SIZE, Config.IMAGE_SIZE)
        with torch.no_grad():
            feats = self.backbone.forward_features(dummy)
        self.feature_dim = feats.shape[1] if feats.ndim == 4 else feats.shape[-1]
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.proj = nn.Sequential(
            nn.Linear(self.feature_dim, 512),
            nn.GELU(),
            nn.Dropout(0.3)
        )

    def forward(self, x):
        feats = self.backbone.forward_features(x)
        if feats.ndim == 4:
            pooled = self.pool(feats).flatten(1)
        else:
            pooled = feats
        return self.proj(pooled)

class ViTFeatureExtractor(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = timm.create_model(Config.VIT_BACKBONE, pretrained=True, num_classes=0)
        for param in self.backbone.parameters():
            param.requires_grad = False
        if hasattr(self.backbone, "blocks"):
            for param in self.backbone.blocks[-2:].parameters():
                param.requires_grad = True

    def forward(self, x):
        return self.backbone(x)

class FastAttention(nn.Module):
    def __init__(self, d_model, num_heads=8):
        super().__init__()
        self.attn = nn.MultiheadAttention(d_model, num_heads,
                                          batch_first=True, dropout=0.1)

    def forward(self, x):
        attn_out, attn_weights = self.attn(x, x, x)
        return attn_out[:, -1:], attn_weights.mean(dim=1)

class FastHybridClassifier(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.cnn = CNNFeatureExtractor()
        self.vit = ViTFeatureExtractor()
        self.lstm = nn.LSTM(512 + 768, 256, 2,
                            batch_first=True, bidirectional=True, dropout=0.2)
        self.attention = FastAttention(512)
        self.classifier = nn.Sequential(
            nn.LayerNorm(512),
            nn.Dropout(0.4),
            nn.Linear(512, 256),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        cnn_feat = self.cnn(x)
        vit_feat = self.vit(x)
        fused = torch.cat([vit_feat, cnn_feat], dim=1).unsqueeze(1)
        seq = [fused + torch.randn_like(fused) * 0.02
               for _ in range(Config.SEQUENCE_LENGTH)]
        seq = torch.cat(seq, dim=1)
        lstm_out, _ = self.lstm(seq)
        attn_out, attn_w = self.attention(lstm_out)
        return self.classifier(attn_out.squeeze(1)), attn_w

# ===================== METRICS & VISUALIZATION =====================

class MetricsVisualizer:
    def __init__(self, class_names, fold_idx=None):
        self.class_names = class_names
        self.fold_idx = fold_idx

    def _prefix(self, name):
        if self.fold_idx is None:
            return name
        base, ext = os.path.splitext(name)
        return f"fold{self.fold_idx}_{base}{ext}"

    def plot_loss_curve(self, train_losses, val_losses):
        plt.figure(figsize=(10, 6))
        plt.plot(train_losses, label='Train Loss', linewidth=3, color='blue')
        plt.plot(val_losses, label='Val Loss', linewidth=3, color='red')
        plt.title('🗺️ Training & Validation Loss', fontsize=16, fontweight='bold', pad=20)
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        fname = self._prefix('1_loss_curve.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_accuracy_curve(self, train_accs, val_accs):
        plt.figure(figsize=(10, 6))
        plt.plot(train_accs, label='Train Acc', linewidth=3, color='green')
        plt.plot(val_accs, label='Val Acc', linewidth=3, color='orange')
        plt.title('📈 Training & Validation Accuracy', fontsize=16, fontweight='bold', pad=20)
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        fname = self._prefix('2_accuracy_curve.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_confusion_matrix(self, y_test, y_pred):
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=self.class_names, yticklabels=self.class_names)
        plt.title('🔍 Confusion Matrix (Test Set)', fontsize=16, fontweight='bold')
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.tight_layout()
        fname = self._prefix('3_confusion_matrix.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_precision_recall(self, y_test, y_pred):
        precision, recall, _, _ = precision_recall_fscore_support(
            y_test, y_pred, average=None)
        x_pos = np.arange(len(self.class_names))
        width = 0.35

        plt.figure(figsize=(12, 6))
        plt.bar(x_pos - width / 2, precision, width,
                label='Precision', alpha=0.8, color='skyblue')
        plt.bar(x_pos + width / 2, recall, width,
                label='Recall', alpha=0.8, color='salmon')
        plt.title('⚖️ Precision & Recall per Class', fontsize=16, fontweight='bold')
        plt.xlabel('Classes')
        plt.ylabel('Score')
        plt.xticks(x_pos, [name[:12] for name in self.class_names], rotation=45)
        plt.legend()
        plt.ylim(0, 1)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        fname = self._prefix('4_precision_recall.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_roc_curves(self, y_test, y_proba):
        plt.figure(figsize=(12, 10))
        y_bin = label_binarize(y_test, classes=range(len(self.class_names)))
        colors = plt.cm.Set1(np.linspace(0, 1, len(self.class_names)))

        for i, color in enumerate(colors):
            if y_bin[:, i].sum() > 0:
                fpr, tpr, _ = roc_curve(y_bin[:, i], y_proba[:, i])
                roc_auc = auc(fpr, tpr)
                plt.plot(fpr, tpr, color=color, lw=2.5,
                         label=f'{self.class_names[i][:15]} (AUC = {roc_auc:.3f})')

        plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Random Classifier')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate', fontsize=12)
        plt.ylabel('True Positive Rate', fontsize=12)
        plt.title('🎯 Multi-Class ROC Curves (GAN-Augmented Test Set)',
                  fontsize=16, fontweight='bold', pad=20)
        plt.legend(loc="lower right", fontsize=11)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        fname = self._prefix('6_roc_curves.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_metrics_summary(self, y_test, y_pred):
        precision_macro, recall_macro, f1_macro, _ = precision_recall_fscore_support(
            y_test, y_pred, average='macro')
        precision_weighted, recall_weighted, f1_weighted, _ = precision_recall_fscore_support(
            y_test, y_pred, average='weighted')
        final_acc = accuracy_score(y_test, y_pred)

        metrics_data = [
            ['Metric', 'Macro Avg', 'Weighted Avg'],
            ['Accuracy', f'{final_acc:.3f}', f'{final_acc:.3f}'],
            ['Precision', f'{precision_macro:.3f}', f'{precision_weighted:.3f}'],
            ['Recall', f'{recall_macro:.3f}', f'{recall_weighted:.3f}'],
            ['F1-Score', f'{f1_macro:.3f}', f'{f1_weighted:.3f}']
        ]

        plt.figure(figsize=(10, 6))
        table = plt.table(cellText=metrics_data[1:],
                          colLabels=metrics_data[0],
                          cellLoc='center', loc='center')
        table.auto_set_font_size(False)
        table.set_fontsize(12)
        table.scale(1, 2.5)
        plt.title('📊 GAN-Augmented Metrics Summary', fontsize=16,
                  fontweight='bold', pad=20)
        plt.axis('off')
        plt.tight_layout()
        fname = self._prefix('7_metrics_summary.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def print_classification_report(self, y_test, y_pred):
        report = classification_report(y_test, y_pred,
                                       target_names=self.class_names,
                                       digits=4)
        print("\n" + "=" * 80)
        print("📋 GAN-AUGMENTED CLASSIFICATION REPORT")
        print("=" * 80)
        print(report)
        fname = self._prefix('classification_report.txt')
        with open(fname, 'w') as f:
            f.write(report)
        print(f"✅ Saved: {fname}")
        print("=" * 80)

# ===================== CLASSIFIER WRAPPER =====================

class FastWheatClassifier:
    def __init__(self, num_classes=5):
        self.model = FastHybridClassifier(num_classes).to(device)
        print(f"Classifier: {sum(p.numel() for p in self.model.parameters())/1e6:.1f}M params")

        self.optimizer = optim.AdamW(self.model.parameters(),
                                     lr=Config.LEARNING_RATE,
                                     weight_decay=Config.WEIGHT_DECAY)
        self.scheduler = CosineAnnealingLR(self.optimizer, T_max=Config.EPOCHS)
        self.criterion = nn.CrossEntropyLoss(label_smoothing=0.1)

        self.class_names = None
        self.train_losses, self.train_accs, self.val_accs, self.val_losses = [], [], [], []

    def update_num_classes(self, num_classes):
        self.model.classifier = nn.Sequential(
            nn.LayerNorm(512),
            nn.Dropout(0.4),
            nn.Linear(512, 256),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_classes)
        ).to(device)
        print(f"🔄 Updated for {num_classes} classes")

    def prepare_data(self, paths, labels, class_names):
        self.image_paths = np.array(paths)
        self.labels = np.array(labels)
        self.class_names = class_names
        num_classes = len(np.unique(self.labels))
        self.update_num_classes(num_classes)
        return self.image_paths, self.labels

    def train(self, train_loader, val_loader):
        best_acc = 0
        best_state = None
        for epoch in range(Config.EPOCHS):
            self.model.train()
            total_loss, correct, total = 0, 0, 0

            pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{Config.EPOCHS}')
            for imgs, lbls in pbar:
                imgs, lbls = imgs.to(device), lbls.to(device)
                self.optimizer.zero_grad()

                if scaler is not None:
                    with torch.cuda.amp.autocast():
                        outs, _ = self.model(imgs)
                        loss = self.criterion(outs, lbls)
                    scaler.scale(loss).backward()
                    scaler.step(self.optimizer)
                    scaler.update()
                else:
                    outs, _ = self.model(imgs)
                    loss = self.criterion(outs, lbls)
                    loss.backward()
                    self.optimizer.step()

                total_loss += loss.item()
                pred = outs.argmax(1)
                correct += (pred == lbls).sum().item()
                total += lbls.size(0)

                pbar.set_postfix({'loss': f'{loss.item():.3f}',
                                  'acc': f'{correct/total:.3f}'})

            train_acc = correct / total
            avg_loss = total_loss / len(train_loader)
            val_acc, val_loss = self.validate(val_loader)

            self.scheduler.step()
            self.train_losses.append(avg_loss)
            self.train_accs.append(train_acc)
            self.val_accs.append(val_acc)
            self.val_losses.append(val_loss)

            if val_acc > best_acc:
                best_acc = val_acc
                best_state = self.model.state_dict()

            print(f'Epoch {epoch+1}: Train {train_acc:.4f} | Val {val_acc:.4f} | Loss {avg_loss:.4f}')

        if best_state is not None:
            self.model.load_state_dict(best_state)
        return best_acc

    @torch.no_grad()
    def validate(self, loader):
        self.model.eval()
        correct, total, loss_sum = 0, 0, 0
        for imgs, lbls in loader:
            imgs, lbls = imgs.to(device), lbls.to(device)
            outs, _ = self.model(imgs)
            loss = self.criterion(outs, lbls)
            loss_sum += loss.item()
            pred = outs.argmax(1)
            correct += (pred == lbls).sum().item()
            total += lbls.size(0)
        return correct / total, loss_sum / len(loader)

    @torch.no_grad()
    def test_full_metrics(self, loader):
        self.model.eval()
        all_preds, all_labels, all_probs = [], [], []
        for imgs, lbls in loader:
            imgs, lbls = imgs.to(device), lbls.to(device)
            outs, _ = self.model(imgs)
            probs = F.softmax(outs, dim=1)
            pred = outs.argmax(1)
            all_preds.extend(pred.cpu().numpy())
            all_labels.extend(lbls.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())

        return np.array(all_labels), np.array(all_preds), np.array(all_probs)

    @torch.no_grad()
    def predict_one_per_class(self, data_dir, output_dir="sample_predictions"):
        os.makedirs(output_dir, exist_ok=True)
        print("\n📸 Generating one prediction per class...")

        orig_paths, orig_labels, _ = detect_dataset_structure(data_dir)

        picked_paths, picked_labels = [], []
        seen_classes = set()
        for path, label in zip(orig_paths, orig_labels):
            if label not in seen_classes:
                picked_paths.append(path)
                picked_labels.append(label)
                seen_classes.add(label)
                if len(seen_classes) == len(self.class_names):
                    break

        _, val_test_transform = get_fast_transforms()
        pred_ds = WheatDiseaseDataset(picked_paths, np.array(picked_labels),
                                      val_test_transform)
        pred_loader = DataLoader(pred_ds, batch_size=1, shuffle=False)

        self.model.eval()

        results = []
        for i, (img, label) in enumerate(pred_loader):
            img = img.to(device)
            label = label.item()

            out, _ = self.model(img)
            prob = F.softmax(out, dim=1)
            pred = prob.argmax(1).item()
            confidence = prob.max().item()

            true_name = self.class_names[label]
            pred_name = self.class_names[pred]

            img_pil = Image.open(picked_paths[i]).convert('RGB').resize((224, 224))
            save_name = f"class_{label}_true-{true_name[:10]}_pred-{pred_name[:10]}_conf-{confidence:.1%}.png"
            save_path = os.path.join(output_dir, save_name)
            img_pil.save(save_path)

            results.append({
                "path": save_path,
                "true_label": label,
                "true_name": true_name,
                "pred_label": pred,
                "pred_name": pred_name,
                "confidence": confidence
            })

        print("\n📸 ONE PREDICTION PER CLASS (Original Dataset):")
        print("-" * 80)
        for r in results:
            status = "✅ CORRECT" if r['true_label'] == r['pred_label'] else "❌ WRONG"
            print(f"{status} | {r['true_name']:<12} → {r['pred_name']:<12} ({r['confidence']:.1%})")
            print(f"           📁 {os.path.basename(r['path'])}")
        print("-" * 80)

        return results

# ===================== HELPERS =====================

def split_data_single(paths, labels):
    train_val, test, train_val_l, test_l = train_test_split(
        paths, labels,
        test_size=Config.TEST_SPLIT_RATIO,
        stratify=labels,
        random_state=42
    )
    train, val, train_l, val_l = train_test_split(
        train_val, train_val_l,
        test_size=0.111,  # ~ 0.1 of original (since 0.9 * 0.111 ≈ 0.1)
        stratify=train_val_l,
        random_state=42
    )
    return train, val, test, train_l, val_l, test_l

def get_weighted_sampler(labels):
    class_counts = np.bincount(labels)
    class_weights = 1.0 / (class_counts + 1e-6)
    sample_weights = class_weights[labels]
    return WeightedRandomSampler(sample_weights, len(sample_weights))

# ===================== MAIN: 3-FOLD CV =====================

def train_fast_gan_3fold(data_dir="/kaggle/input/wheat-disease"):
    print(f"🎯 Starting 3-FOLD GAN-ENHANCED training: {data_dir}")

    # Build GAN-augmented dataset once
    all_paths, all_labels, class_names = prepare_gan_augmented_data_simple(data_dir)
    all_paths = np.array(all_paths)
    all_labels = np.array(all_labels)

    skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
    fold_accuracies = []

    for fold_idx, (train_val_idx, test_idx) in enumerate(skf.split(all_paths, all_labels), start=1):
        print("\n" + "=" * 80)
        print(f"🧩 FOLD {fold_idx}/3")
        print("=" * 80)

        paths_train_val = all_paths[train_val_idx]
        labels_train_val = all_labels[train_val_idx]
        paths_test = all_paths[test_idx]
        labels_test = all_labels[test_idx]

        train_p, val_p, _, train_l, val_l, _ = split_data_single(
            paths_train_val, labels_train_val
        )
        test_p = paths_test
        test_l = labels_test

        print(f"📊 Fold {fold_idx} splits (Real + Synthetic):")
        print(f"   Train: {len(train_p)} | Val: {len(val_p)} | Test: {len(test_p)}")

        train_transform, val_test_transform = get_fast_transforms()
        train_ds = WheatDiseaseDataset(train_p, train_l, train_transform)
        val_ds = WheatDiseaseDataset(val_p, val_l, val_test_transform)
        test_ds = WheatDiseaseDataset(test_p, test_l, val_test_transform)

        sampler = get_weighted_sampler(train_l)
        train_loader = DataLoader(
            train_ds, Config.BATCH_SIZE, sampler=sampler,
            num_workers=Config.NUM_WORKERS, pin_memory=True
        )
        val_loader = DataLoader(
            val_ds, Config.BATCH_SIZE, shuffle=False,
            num_workers=Config.NUM_WORKERS, pin_memory=True
        )
        test_loader = DataLoader(
            test_ds, Config.BATCH_SIZE, shuffle=False,
            num_workers=Config.NUM_WORKERS, pin_memory=True
        )

        system = FastWheatClassifier(num_classes=len(class_names))
        system.prepare_data(all_paths, all_labels, class_names)

        # Reset optimizer & scheduler
        system.optimizer = optim.AdamW(system.model.parameters(),
                                       lr=Config.LEARNING_RATE,
                                       weight_decay=Config.WEIGHT_DECAY)
        system.scheduler = CosineAnnealingLR(system.optimizer, T_max=Config.EPOCHS)
        system.train_losses, system.train_accs, system.val_accs, system.val_losses = [], [], [], []

        print(f"🚀 CLASSIFIER TRAINING (GAN Data) - Fold {fold_idx}")
        best_val = system.train(train_loader, val_loader)

        print(f"\n📊 FULL EVALUATION + VISUALIZATION - Fold {fold_idx}")
        y_test, y_pred, y_proba = system.test_full_metrics(test_loader)

        visualizer = MetricsVisualizer(system.class_names, fold_idx=fold_idx)
        visualizer.plot_loss_curve(system.train_losses, system.val_losses)
        visualizer.plot_accuracy_curve(system.train_accs, system.val_accs)
        visualizer.plot_confusion_matrix(y_test, y_pred)
        visualizer.plot_precision_recall(y_test, y_pred)
        visualizer.plot_roc_curves(y_test, y_proba)
        visualizer.plot_metrics_summary(y_test, y_pred)
        visualizer.print_classification_report(y_test, y_pred)

        final_acc = accuracy_score(y_test, y_pred)
        fold_accuracies.append(final_acc)
        print(f"\n🎉 FOLD {fold_idx} TEST ACCURACY: {final_acc:.4f} ({final_acc:.1%})")

        # Optional: per-fold one-per-class predictions
        system.predict_one_per_class(
            data_dir, output_dir=f"sample_predictions_fold{fold_idx}"
        )

    print("\n" + "#" * 80)
    print("📦 3-FOLD RESULTS")
    for i, acc in enumerate(fold_accuracies, start=1):
        print(f"   Fold {i}: {acc:.4f} ({acc:.1%})")
    mean_acc = np.mean(fold_accuracies)
    std_acc = np.std(fold_accuracies)
    print(f"   Mean Acc: {mean_acc:.4f} ± {std_acc:.4f}")
    print("#" * 80 + "\n")

    return fold_accuracies, mean_acc, std_acc

if __name__ == "__main__":
    fold_accs, mean_acc, std_acc = train_fast_gan_3fold(
        "/kaggle/input/wheat-disease"
    )
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
import numpy as np
from PIL import Image
import os
import random
import torchvision.transforms as transforms
import timm
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, \
    precision_recall_fscore_support, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import warnings
from torch.optim.lr_scheduler import CosineAnnealingLR
import albumentations as A
from albumentations.pytorch import ToTensorV2
from sklearn.preprocessing import label_binarize

warnings.filterwarnings("ignore")

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
scaler = torch.cuda.amp.GradScaler() if device.type == 'cuda' else None
print(f"Using device: {device}")

class Config:
    IMAGE_SIZE = 224
    BATCH_SIZE = 64
    NUM_WORKERS = 8

    # Binary dataset
    NUM_CLASSES = 2

    # Reduced epochs for classifier
    EPOCHS = 15
    LEARNING_RATE = 3e-4
    WEIGHT_DECAY = 1e-4

    # Reduced GAN epochs
    GAN_EPOCHS = 20
    GAN_BATCH_SIZE = 32
    GAN_LR = 2e-4
    GAN_FEATURE_MATCHING = True
    GAN_AUGMENTATION_RATIO = 1

    TRAIN_SPLIT_RATIO = 0.8
    TEST_SPLIT_RATIO = 0.1
    VIT_BACKBONE = 'vit_base_patch16_224'
    CNN_BACKBONE = 'efficientnet_b2'
    LSTM_HIDDEN_SIZE = 256
    LSTM_NUM_LAYERS = 2
    LSTM_DROPOUT = 0.3
    SEQUENCE_LENGTH = 6

# ===================== GAN MODULES =====================

class ConditionalGenerator(nn.Module):
    def __init__(self, num_classes, latent_dim=100, img_channels=3):
        super().__init__()
        self.label_emb = nn.Embedding(num_classes, 64)

        self.model = nn.Sequential(
            nn.Linear(latent_dim + 64, 256 * 4 * 4),
            nn.ReLU(),
            nn.Unflatten(1, (256, 4, 4)),
            nn.ConvTranspose2d(256, 128, 4, 2, 1), nn.BatchNorm2d(128), nn.ReLU(),
            nn.ConvTranspose2d(128, 64, 4, 2, 1), nn.BatchNorm2d(64), nn.ReLU(),
            nn.ConvTranspose2d(64, 32, 4, 2, 1), nn.BatchNorm2d(32), nn.ReLU(),
            nn.ConvTranspose2d(32, img_channels, 4, 2, 1), nn.Tanh()
        )

    def forward(self, z, labels):
        c = self.label_emb(labels)
        x = torch.cat([z, c], dim=1)
        img = self.model(x)
        return img

class ConditionalDiscriminator(nn.Module):
    def __init__(self, num_classes, img_channels=3):
        super().__init__()
        self.label_emb = nn.Embedding(num_classes, 64)

        self.model = nn.Sequential(
            nn.Conv2d(img_channels + 64, 64, 4, 2, 1), nn.LeakyReLU(0.2),
            nn.Conv2d(64, 128, 4, 2, 1), nn.BatchNorm2d(128), nn.LeakyReLU(0.2),
            nn.Conv2d(128, 256, 4, 2, 1), nn.BatchNorm2d(256), nn.LeakyReLU(0.2),
            nn.Conv2d(256, 512, 4, 2, 1), nn.BatchNorm2d(512), nn.LeakyReLU(0.2),
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),
            nn.Linear(512, 1)
        )

    def forward(self, img, labels):
        c = self.label_emb(labels).view(labels.size(0), 64, 1, 1)
        c = c.repeat(1, 1, img.size(2), img.size(3))
        x = torch.cat([img, c], dim=1)
        return self.model(x)

class WheatDiseaseGAN:
    def __init__(self, num_classes):
        self.num_classes = num_classes
        self.generator = ConditionalGenerator(num_classes).to(device)
        self.discriminator = ConditionalDiscriminator(num_classes).to(device)

        self.g_optimizer = optim.Adam(self.generator.parameters(), lr=Config.GAN_LR, betas=(0.5, 0.999))
        self.d_optimizer = optim.Adam(self.discriminator.parameters(), lr=Config.GAN_LR, betas=(0.5, 0.999))

        print(f"🆕 GAN - G: {sum(p.numel() for p in self.generator.parameters())/1e6:.1f}M, "
              f"D: {sum(p.numel() for p in self.discriminator.parameters())/1e6:.1f}M")

    def train_gan(self, dataloader, epochs=Config.GAN_EPOCHS):
        print("🚀 Training Conditional GAN...")
        criterion = nn.BCEWithLogitsLoss()

        for epoch in range(epochs):
            g_loss_epoch, d_loss_epoch = 0, 0

            pbar = tqdm(dataloader, desc=f'GAN E{epoch+1}', leave=False)
            for real_imgs, labels in pbar:
                batch_size = real_imgs.size(0)
                real_imgs = real_imgs.to(device)
                labels = labels.to(device)

                real_labels = torch.ones(batch_size, 1).to(device)
                fake_labels = torch.zeros(batch_size, 1).to(device)

                # Train discriminator
                self.d_optimizer.zero_grad()
                real_pred = self.discriminator(real_imgs, labels)
                d_real_loss = criterion(real_pred, real_labels)

                z = torch.randn(batch_size, 100).to(device)
                fake_imgs = self.generator(z, labels)
                fake_pred = self.discriminator(fake_imgs.detach(), labels)
                d_fake_loss = criterion(fake_pred, fake_labels)

                d_loss = (d_real_loss + d_fake_loss) / 2
                d_loss.backward()
                self.d_optimizer.step()

                # Train generator
                self.g_optimizer.zero_grad()
                fake_pred = self.discriminator(fake_imgs, labels)
                g_loss = criterion(fake_pred, real_labels)

                if Config.GAN_FEATURE_MATCHING:
                    feat_real = torch.mean(real_imgs, dim=[2, 3], keepdim=True)
                    feat_fake = torch.mean(fake_imgs, dim=[2, 3], keepdim=True)
                    fm_loss = F.mse_loss(feat_fake, feat_real)
                    g_loss += 0.1 * fm_loss

                g_loss.backward()
                self.g_optimizer.step()

                g_loss_epoch += g_loss.item()
                d_loss_epoch += d_loss.item()
                pbar.set_postfix({'D': f'{d_loss.item():.3f}', 'G': f'{g_loss.item():.3f}'})

            if epoch % 5 == 0 or epoch == epochs - 1:
                print(f'GAN E[{epoch+1}/{epochs}] D:{d_loss_epoch/len(dataloader):.3f} '
                      f'G:{g_loss_epoch/len(dataloader):.3f}')

        print("✅ GAN trained!")

    @torch.no_grad()
    def generate_synthetic_data(self, num_samples_per_class):
        print("🎨 Generating synthetic images...")
        synthetic_paths, synthetic_labels = [], []
        os.makedirs("synthetic_data", exist_ok=True)

        self.generator.eval()
        with torch.no_grad():
            for class_idx in range(self.num_classes):
                labels = torch.full((num_samples_per_class,), class_idx,
                                    dtype=torch.long).to(device)
                z = torch.randn(num_samples_per_class, 100).to(device)
                fake_imgs = self.generator(z, labels)

                class_dir = f"synthetic_data/class_{class_idx}"
                os.makedirs(class_dir, exist_ok=True)

                for i, img in enumerate(fake_imgs):
                    img = (img * 0.5 + 0.5).clamp(0, 1)
                    img_pil = transforms.ToPILImage()(img.cpu())
                    path = os.path.join(class_dir, f"syn_{i:04d}.png")
                    img_pil.save(path)
                    synthetic_paths.append(path)
                    synthetic_labels.append(class_idx)

        print(f"✅ Generated {len(synthetic_paths)} images")
        return synthetic_paths, np.array(synthetic_labels)

# ===================== DATA & TRANSFORMS =====================

def detect_dataset_structure(data_dir):
    class_names, image_paths, labels = [], [], []
    print(f"🔍 Scanning: {data_dir}")

    for root, dirs, files in os.walk(data_dir):
        image_files = [f for f in files if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        if not image_files:
            continue

        class_name = os.path.basename(root)
        if class_name not in class_names:
            class_names.append(class_name)
            class_names.sort()

        class_idx = class_names.index(class_name)
        for img_file in image_files:
            image_paths.append(os.path.join(root, img_file))
            labels.append(class_idx)

    print(f"✅ Found {len(class_names)} classes: {class_names}")
    print(f"📊 Total images: {len(image_paths)}")
    return image_paths, np.array(labels), class_names

def get_fast_transforms():
    train_transform = A.Compose([
        A.Resize(Config.IMAGE_SIZE, Config.IMAGE_SIZE),
        A.HorizontalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.RandomBrightnessContrast(p=0.3),
        A.Normalize(mean=[0.485, 0.456, 0.406],
                    std=[0.229, 0.224, 0.225]),
        ToTensorV2()
    ])
    val_test_transform = A.Compose([
        A.Resize(Config.IMAGE_SIZE, Config.IMAGE_SIZE),
        A.Normalize(mean=[0.485, 0.456, 0.406],
                    std=[0.229, 0.224, 0.225]),
        ToTensorV2()
    ])
    return train_transform, val_test_transform

class WheatDiseaseDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = list(image_paths)
        self.labels = np.array(labels)
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        try:
            image = Image.open(self.image_paths[idx]).convert('RGB')
            image = np.array(image)
            label = int(self.labels[idx])
            if self.transform:
                augmented = self.transform(image=image)
                image = augmented['image']
            return image, label
        except:
            dummy = np.full((Config.IMAGE_SIZE, Config.IMAGE_SIZE, 3),
                            128, dtype=np.uint8)
            return torch.tensor(dummy.transpose(2, 0, 1) / 255.0), 0

def prepare_gan_augmented_data_simple(data_dir):
    orig_paths, orig_labels, class_names = detect_dataset_structure(data_dir)
    if len(orig_paths) == 0:
        raise ValueError("No images found!")

    _, val_test_transform = get_fast_transforms()

    gan_dataset = WheatDiseaseDataset(orig_paths, orig_labels, val_test_transform)
    gan_loader = DataLoader(
        gan_dataset,
        batch_size=Config.GAN_BATCH_SIZE,
        shuffle=True,
        num_workers=4,
        pin_memory=True
    )

    gan_trainer = WheatDiseaseGAN(len(class_names))
    gan_trainer.train_gan(gan_loader, epochs=Config.GAN_EPOCHS)

    orig_counts = np.bincount(orig_labels)
    target_per_class = min(int(np.min(orig_counts) * Config.GAN_AUGMENTATION_RATIO), 400)
    syn_paths, syn_labels = gan_trainer.generate_synthetic_data(target_per_class)

    all_paths = orig_paths + syn_paths
    all_labels = np.concatenate([orig_labels, syn_labels])

    print(f"📊 GAN Augmented: {len(orig_paths)} → {len(all_paths)} images")
    return all_paths, all_labels, class_names

# ===================== MODEL (CNN + ViT + LSTM) =====================

class CNNFeatureExtractor(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = timm.create_model(Config.CNN_BACKBONE, pretrained=True, num_classes=0)
        for param in self.backbone.parameters():
            param.requires_grad = False
        if hasattr(self.backbone, "blocks"):
            for param in self.backbone.blocks[-2:].parameters():
                param.requires_grad = True

        dummy = torch.randn(1, 3, Config.IMAGE_SIZE, Config.IMAGE_SIZE)
        with torch.no_grad():
            feats = self.backbone.forward_features(dummy)
        self.feature_dim = feats.shape[1] if feats.ndim == 4 else feats.shape[-1]
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.proj = nn.Sequential(
            nn.Linear(self.feature_dim, 512),
            nn.GELU(),
            nn.Dropout(0.3)
        )

    def forward(self, x):
        feats = self.backbone.forward_features(x)
        if feats.ndim == 4:
            pooled = self.pool(feats).flatten(1)
        else:
            pooled = feats
        return self.proj(pooled)

class ViTFeatureExtractor(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = timm.create_model(Config.VIT_BACKBONE, pretrained=True, num_classes=0)
        for param in self.backbone.parameters():
            param.requires_grad = False
        if hasattr(self.backbone, "blocks"):
            for param in self.backbone.blocks[-2:].parameters():
                param.requires_grad = True

    def forward(self, x):
        return self.backbone(x)

class FastAttention(nn.Module):
    def __init__(self, d_model, num_heads=8):
        super().__init__()
        self.attn = nn.MultiheadAttention(d_model, num_heads,
                                          batch_first=True, dropout=0.1)

    def forward(self, x):
        attn_out, attn_weights = self.attn(x, x, x)
        return attn_out[:, -1:], attn_weights.mean(dim=1)

class FastHybridClassifier(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.cnn = CNNFeatureExtractor()
        self.vit = ViTFeatureExtractor()
        self.lstm = nn.LSTM(512 + 768, 256, 2,
                            batch_first=True, bidirectional=True, dropout=0.2)
        self.attention = FastAttention(512)
        self.classifier = nn.Sequential(
            nn.LayerNorm(512),
            nn.Dropout(0.4),
            nn.Linear(512, 256),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        cnn_feat = self.cnn(x)
        vit_feat = self.vit(x)
        fused = torch.cat([vit_feat, cnn_feat], dim=1).unsqueeze(1)
        seq = [fused + torch.randn_like(fused) * 0.02
               for _ in range(Config.SEQUENCE_LENGTH)]
        seq = torch.cat(seq, dim=1)
        lstm_out, _ = self.lstm(seq)
        attn_out, attn_w = self.attention(lstm_out)
        return self.classifier(attn_out.squeeze(1)), attn_w

# ===================== METRICS & VISUALIZATION =====================

class MetricsVisualizer:
    def __init__(self, class_names, fold_idx=None):
        self.class_names = class_names
        self.fold_idx = fold_idx

    def _prefix(self, name):
        if self.fold_idx is None:
            return name
        base, ext = os.path.splitext(name)
        return f"fold{self.fold_idx}_{base}{ext}"

    def plot_loss_curve(self, train_losses, val_losses):
        plt.figure(figsize=(10, 6))
        plt.plot(train_losses, label='Train Loss', linewidth=3, color='blue')
        plt.plot(val_losses, label='Val Loss', linewidth=3, color='red')
        plt.title('🗺️ Training & Validation Loss', fontsize=16, fontweight='bold', pad=20)
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        fname = self._prefix('1_loss_curve.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_accuracy_curve(self, train_accs, val_accs):
        plt.figure(figsize=(10, 6))
        plt.plot(train_accs, label='Train Acc', linewidth=3, color='green')
        plt.plot(val_accs, label='Val Acc', linewidth=3, color='orange')
        plt.title('📈 Training & Validation Accuracy', fontsize=16, fontweight='bold', pad=20)
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        fname = self._prefix('2_accuracy_curve.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_confusion_matrix(self, y_test, y_pred):
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=self.class_names, yticklabels=self.class_names)
        plt.title('🔍 Confusion Matrix (Test Set)', fontsize=16, fontweight='bold')
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.tight_layout()
        fname = self._prefix('3_confusion_matrix.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_precision_recall(self, y_test, y_pred):
        precision, recall, _, _ = precision_recall_fscore_support(
            y_test, y_pred, average=None)
        x_pos = np.arange(len(self.class_names))
        width = 0.35

        plt.figure(figsize=(12, 6))
        plt.bar(x_pos - width / 2, precision, width,
                label='Precision', alpha=0.8, color='skyblue')
        plt.bar(x_pos + width / 2, recall, width,
                label='Recall', alpha=0.8, color='salmon')
        plt.title('⚖️ Precision & Recall per Class', fontsize=16, fontweight='bold')
        plt.xlabel('Classes')
        plt.ylabel('Score')
        plt.xticks(x_pos, [name[:12] for name in self.class_names], rotation=45)
        plt.legend()
        plt.ylim(0, 1)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        fname = self._prefix('4_precision_recall.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_roc_curves(self, y_test, y_proba):
        plt.figure(figsize=(12, 10))
        n_classes = len(self.class_names)

        if n_classes == 2:
            # Binary ROC: use positive class probability (column 1)
            y_score = y_proba[:, 1]
            fpr, tpr, _ = roc_curve(y_test, y_score)
            roc_auc = auc(fpr, tpr)
            plt.plot(
                fpr,
                tpr,
                color='darkorange',
                lw=2.5,
                label=f'{self.class_names[1]} vs {self.class_names[0]} (AUC = {roc_auc:.3f})'
            )
        else:
            # Multi-class (OvR)
            y_bin = label_binarize(y_test, classes=range(n_classes))
            colors = plt.cm.Set1(np.linspace(0, 1, n_classes))

            for i, color in enumerate(colors):
                if y_bin[:, i].sum() > 0:
                    fpr, tpr, _ = roc_curve(y_bin[:, i], y_proba[:, i])
                    roc_auc = auc(fpr, tpr)
                    plt.plot(
                        fpr,
                        tpr,
                        color=color,
                        lw=2.5,
                        label=f'{self.class_names[i][:15]} (AUC = {roc_auc:.3f})'
                    )

        plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Random Classifier')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate', fontsize=12)
        plt.ylabel('True Positive Rate', fontsize=12)
        plt.title('🎯 ROC Curve', fontsize=16, fontweight='bold', pad=20)
        plt.legend(loc="lower right", fontsize=11)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        fname = self._prefix('6_roc_curves.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def plot_metrics_summary(self, y_test, y_pred):
        precision_macro, recall_macro, f1_macro, _ = precision_recall_fscore_support(
            y_test, y_pred, average='macro')
        precision_weighted, recall_weighted, f1_weighted, _ = precision_recall_fscore_support(
            y_test, y_pred, average='weighted')
        final_acc = accuracy_score(y_test, y_pred)

        metrics_data = [
            ['Metric', 'Macro Avg', 'Weighted Avg'],
            ['Accuracy', f'{final_acc:.3f}', f'{final_acc:.3f}'],
            ['Precision', f'{precision_macro:.3f}', f'{precision_weighted:.3f}'],
            ['Recall', f'{recall_macro:.3f}', f'{recall_weighted:.3f}'],
            ['F1-Score', f'{f1_macro:.3f}', f'{f1_weighted:.3f}']
        ]

        plt.figure(figsize=(10, 6))
        table = plt.table(cellText=metrics_data[1:],
                          colLabels=metrics_data[0],
                          cellLoc='center', loc='center')
        table.auto_set_font_size(False)
        table.set_fontsize(12)
        table.scale(1, 2.5)
        plt.title('📊 GAN-Augmented Metrics Summary', fontsize=16,
                  fontweight='bold', pad=20)
        plt.axis('off')
        plt.tight_layout()
        fname = self._prefix('7_metrics_summary.png')
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Saved: {fname}")

    def print_classification_report(self, y_test, y_pred):
        report = classification_report(y_test, y_pred,
                                       target_names=self.class_names,
                                       digits=4)
        print("\n" + "=" * 80)
        print("📋 GAN-AUGMENTED CLASSIFICATION REPORT")
        print("=" * 80)
        print(report)
        fname = self._prefix('classification_report.txt')
        with open(fname, 'w') as f:
            f.write(report)
        print(f"✅ Saved: {fname}")
        print("=" * 80)

# ===================== CLASSIFIER WRAPPER =====================

class FastWheatClassifier:
    def __init__(self, num_classes=Config.NUM_CLASSES):
        self.model = FastHybridClassifier(num_classes).to(device)
        print(f"Classifier: {sum(p.numel() for p in self.model.parameters())/1e6:.1f}M params")

        self.optimizer = optim.AdamW(self.model.parameters(),
                                     lr=Config.LEARNING_RATE,
                                     weight_decay=Config.WEIGHT_DECAY)
        self.scheduler = CosineAnnealingLR(self.optimizer, T_max=Config.EPOCHS)
        self.criterion = nn.CrossEntropyLoss(label_smoothing=0.1)

        self.class_names = None
        self.train_losses, self.train_accs, self.val_accs, self.val_losses = [], [], [], []

    def update_num_classes(self, num_classes):
        self.model.classifier = nn.Sequential(
            nn.LayerNorm(512),
            nn.Dropout(0.4),
            nn.Linear(512, 256),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_classes)
        ).to(device)
        print(f"🔄 Updated for {num_classes} classes")

    def prepare_data(self, paths, labels, class_names):
        self.image_paths = np.array(paths)
        self.labels = np.array(labels)
        self.class_names = class_names
        num_classes = len(np.unique(self.labels))
        self.update_num_classes(num_classes)
        return self.image_paths, self.labels

    def train(self, train_loader, val_loader):
        best_acc = 0
        best_state = None
        for epoch in range(Config.EPOCHS):
            self.model.train()
            total_loss, correct, total = 0, 0, 0

            pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{Config.EPOCHS}')
            for imgs, lbls in pbar:
                imgs, lbls = imgs.to(device), lbls.to(device)
                self.optimizer.zero_grad()

                if scaler is not None:
                    with torch.cuda.amp.autocast():
                        outs, _ = self.model(imgs)
                        loss = self.criterion(outs, lbls)
                    scaler.scale(loss).backward()
                    scaler.step(self.optimizer)
                    scaler.update()
                else:
                    outs, _ = self.model(imgs)
                    loss = self.criterion(outs, lbls)
                    loss.backward()
                    self.optimizer.step()

                total_loss += loss.item()
                pred = outs.argmax(1)
                correct += (pred == lbls).sum().item()
                total += lbls.size(0)

                pbar.set_postfix({'loss': f'{loss.item():.3f}',
                                  'acc': f'{correct/total:.3f}'})

            train_acc = correct / total
            avg_loss = total_loss / len(train_loader)
            val_acc, val_loss = self.validate(val_loader)

            self.scheduler.step()
            self.train_losses.append(avg_loss)
            self.train_accs.append(train_acc)
            self.val_accs.append(val_acc)
            self.val_losses.append(val_loss)

            if val_acc > best_acc:
                best_acc = val_acc
                best_state = self.model.state_dict()

            print(f'Epoch {epoch+1}: Train {train_acc:.4f} | Val {val_acc:.4f} | Loss {avg_loss:.4f}')

        if best_state is not None:
            self.model.load_state_dict(best_state)
        return best_acc

    @torch.no_grad()
    def validate(self, loader):
        self.model.eval()
        correct, total, loss_sum = 0, 0, 0
        for imgs, lbls in loader:
            imgs, lbls = imgs.to(device), lbls.to(device)
            outs, _ = self.model(imgs)
            loss = self.criterion(outs, lbls)
            loss_sum += loss.item()
            pred = outs.argmax(1)
            correct += (pred == lbls).sum().item()
            total += lbls.size(0)
        return correct / total, loss_sum / len(loader)

    @torch.no_grad()
    def test_full_metrics(self, loader):
        self.model.eval()
        all_preds, all_labels, all_probs = [], [], []
        for imgs, lbls in loader:
            imgs, lbls = imgs.to(device), lbls.to(device)
            outs, _ = self.model(imgs)
            probs = F.softmax(outs, dim=1)
            pred = outs.argmax(1)
            all_preds.extend(pred.cpu().numpy())
            all_labels.extend(lbls.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())

        return np.array(all_labels), np.array(all_preds), np.array(all_probs)

    @torch.no_grad()
    def predict_one_per_class(self, data_dir, output_dir="sample_predictions"):
        os.makedirs(output_dir, exist_ok=True)
        print("\n📸 Generating one prediction per class...")

        orig_paths, orig_labels, _ = detect_dataset_structure(data_dir)

        picked_paths, picked_labels = [], []
        seen_classes = set()
        for path, label in zip(orig_paths, orig_labels):
            if label not in seen_classes:
                picked_paths.append(path)
                picked_labels.append(label)
                seen_classes.add(label)
                if len(seen_classes) == len(self.class_names):
                    break

        _, val_test_transform = get_fast_transforms()
        pred_ds = WheatDiseaseDataset(picked_paths, np.array(picked_labels),
                                      val_test_transform)
        pred_loader = DataLoader(pred_ds, batch_size=1, shuffle=False)

        self.model.eval()

        results = []
        for i, (img, label) in enumerate(pred_loader):
            img = img.to(device)
            label = label.item()

            out, _ = self.model(img)
            prob = F.softmax(out, dim=1)
            pred = prob.argmax(1).item()
            confidence = prob.max().item()

            true_name = self.class_names[label]
            pred_name = self.class_names[pred]

            img_pil = Image.open(picked_paths[i]).convert('RGB').resize((224, 224))
            save_name = f"class_{label}_true-{true_name[:10]}_pred-{pred_name[:10]}_conf-{confidence:.1%}.png"
            save_path = os.path.join(output_dir, save_name)
            img_pil.save(save_path)

            results.append({
                "path": save_path,
                "true_label": label,
                "true_name": true_name,
                "pred_label": pred,
                "pred_name": pred_name,
                "confidence": confidence
            })

        print("\n📸 ONE PREDICTION PER CLASS (Original Dataset):")
        print("-" * 80)
        for r in results:
            status = "✅ CORRECT" if r['true_label'] == r['pred_label'] else "❌ WRONG"
            print(f"{status} | {r['true_name']:<12} → {r['pred_name']:<12} ({r['confidence']:.1%})")
            print(f"           📁 {os.path.basename(r['path'])}")
        print("-" * 80)

        return results

# ===================== HELPERS =====================

def split_data_single(paths, labels):
    train_val, test, train_val_l, test_l = train_test_split(
        paths, labels,
        test_size=Config.TEST_SPLIT_RATIO,
        stratify=labels,
        random_state=42
    )
    train, val, train_l, val_l = train_test_split(
        train_val, train_val_l,
        test_size=0.111,
        stratify=train_val_l,
        random_state=42
    )
    return train, val, test, train_l, val_l, test_l

def get_weighted_sampler(labels):
    class_counts = np.bincount(labels)
    class_weights = 1.0 / (class_counts + 1e-6)
    sample_weights = class_weights[labels]
    return WeightedRandomSampler(sample_weights, len(sample_weights))

# ===================== MAIN: 3-FOLD CV =====================

def train_fast_gan_3fold(data_dir):
    print(f"🎯 Starting 3-FOLD GAN-ENHANCED training: {data_dir}")

    all_paths, all_labels, class_names = prepare_gan_augmented_data_simple(data_dir)
    all_paths = np.array(all_paths)
    all_labels = np.array(all_labels)

    skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
    fold_accuracies = []

    for fold_idx, (train_val_idx, test_idx) in enumerate(skf.split(all_paths, all_labels), start=1):
        print("\n" + "=" * 80)
        print(f"🧩 FOLD {fold_idx}/3")
        print("=" * 80)

        paths_train_val = all_paths[train_val_idx]
        labels_train_val = all_labels[train_val_idx]
        paths_test = all_paths[test_idx]
        labels_test = all_labels[test_idx]

        train_p, val_p, _, train_l, val_l, _ = split_data_single(
            paths_train_val, labels_train_val
        )
        test_p = paths_test
        test_l = labels_test

        print(f"📊 Fold {fold_idx} splits (Real + Synthetic):")
        print(f"   Train: {len(train_p)} | Val: {len(val_p)} | Test: {len(test_p)}")

        train_transform, val_test_transform = get_fast_transforms()
        train_ds = WheatDiseaseDataset(train_p, train_l, train_transform)
        val_ds = WheatDiseaseDataset(val_p, val_l, val_test_transform)
        test_ds = WheatDiseaseDataset(test_p, test_l, val_test_transform)

        sampler = get_weighted_sampler(train_l)
        train_loader = DataLoader(
            train_ds, Config.BATCH_SIZE, sampler=sampler,
            num_workers=Config.NUM_WORKERS, pin_memory=True
        )
        val_loader = DataLoader(
            val_ds, Config.BATCH_SIZE, shuffle=False,
            num_workers=Config.NUM_WORKERS, pin_memory=True
        )
        test_loader = DataLoader(
            test_ds, Config.BATCH_SIZE, shuffle=False,
            num_workers=Config.NUM_WORKERS, pin_memory=True
        )

        system = FastWheatClassifier(num_classes=len(class_names))
        system.prepare_data(all_paths, all_labels, class_names)

        system.optimizer = optim.AdamW(system.model.parameters(),
                                       lr=Config.LEARNING_RATE,
                                       weight_decay=Config.WEIGHT_DECAY)
        system.scheduler = CosineAnnealingLR(system.optimizer, T_max=Config.EPOCHS)
        system.train_losses, system.train_accs, system.val_accs, system.val_losses = [], [], [], []

        print(f"🚀 CLASSIFIER TRAINING (GAN Data) - Fold {fold_idx}")
        best_val = system.train(train_loader, val_loader)

        print(f"\n📊 FULL EVALUATION + VISUALIZATION - Fold {fold_idx}")
        y_test, y_pred, y_proba = system.test_full_metrics(test_loader)

        visualizer = MetricsVisualizer(system.class_names, fold_idx=fold_idx)
        visualizer.plot_loss_curve(system.train_losses, system.val_losses)
        visualizer.plot_accuracy_curve(system.train_accs, system.val_accs)
        visualizer.plot_confusion_matrix(y_test, y_pred)
        visualizer.plot_precision_recall(y_test, y_pred)
        visualizer.plot_roc_curves(y_test, y_proba)
        visualizer.plot_metrics_summary(y_test, y_pred)
        visualizer.print_classification_report(y_test, y_pred)

        final_acc = accuracy_score(y_test, y_pred)
        fold_accuracies.append(final_acc)
        print(f"\n🎉 FOLD {fold_idx} TEST ACCURACY: {final_acc:.4f} ({final_acc:.1%})")

        system.predict_one_per_class(
            data_dir, output_dir=f"sample_predictions_fold{fold_idx}"
        )

    print("\n" + "#" * 80)
    print("📦 3-FOLD RESULTS")
    for i, acc in enumerate(fold_accuracies, start=1):
        print(f"   Fold {i}: {acc:.4f} ({acc:.1%})")
    mean_acc = np.mean(fold_accuracies)
    std_acc = np.std(fold_accuracies)
    print(f"   Mean Acc: {mean_acc:.4f} ± {std_acc:.4f}")
    print("#" * 80 + "\n")

    return fold_accuracies, mean_acc, std_acc

if __name__ == "__main__":
    # Replace this with your actual 2-class dataset root
    fold_accs, mean_acc, std_acc = train_fast_gan_3fold(
        "/kaggle/input/new-wheat-leaf-disease-dataset/Wheat Leaf Disease Dataset"
    )
```
